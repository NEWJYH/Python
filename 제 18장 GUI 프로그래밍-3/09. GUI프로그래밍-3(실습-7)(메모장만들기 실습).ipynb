```python
# memo

# 메모장 만들기
# GUI 만들기

from tkinter import *

window = Tk()
window.title("메모장")
window.geometry("400x400")

text = Text(window) # text위젯을 생성함
sb = Scrollbar(text) # 텍스트 위젯의 스크롤바를 생성함

# 수직으로 생성하겠다는 의미
sb.config(command=text.yview())
sb.pack(side=RIGHT, fill=Y)

# 메모장의 기본크기를 400x400 으로 설정했지만
# 필요하다면 메모장의 크기를 확대를 시킬수도 있다.
# 이때 Text위젯이 자동으로 확대가 되기 위해서 아래와 같은 코드를 작성함

# 줄을 전체 크기로 설정함
# weight 는 상대적인 크기를 나타내는 매개변수이며 1로 지정을 하면
# 바로 전체화면의 크기에 맞추어서 확장이 자동으로 이루어진다.
window.grid_rowconfigure(0, weight=1)
# 열을 전체 크기로 설정함
window.grid_columnconfigure(0, weight=1)

# 텍스트 위젯이 좌우 상하를 다 채우도록 고정
text.grid(sticky=N+E+S+W) 
   
    
    
    
# 메뉴 만들기
mb = Menu(window)
# 파일 메뉴를 만드는 코드
fi = Menu(mb,tearoff=False)
fi.add_command(label="새파일")
fi.add_command(label="열기")
fi.add_command(label="저장")
# 분리선 추가
fi.add_separator()
fi.add_command(label="종료")
# 파일 메뉴를 메뉴바에 붙이기
mb.add_cascade(label="파일",menu=fi)
window.config(menu=mb)




# 편집 메뉴를 만드는 코드
e = Menu(mb, tearoff=False) # 점선 없애기
e.add_command(label="잘라내기")
e.add_command(label="복사")
e.add_command(label="붙이기")
e.add_command(label="삭제")
# 편집 메뉴를 메뉴바에 붙이기
mb.add_cascade(label="편집",menu=e)





# 도움말 메뉴를 만드는 코드
h = Menu(mb,tearoff=False)
h.add_command(label="메모장 정보")
# 도움말 메뉴를 메뉴바에 붙이기
mb.add_cascade(label="도움말",menu=h)

window.mainloop()
```
