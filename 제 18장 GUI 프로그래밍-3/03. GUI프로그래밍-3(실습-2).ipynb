```python
# 마우스 모션에 대한 실습
from tkinter import *

# 이벤트 처리 함수 정의
# 마우스가 움직이면 움직인 좌표값을 찍는다.
def motion(event):
    print(f"마우스의 위치 : {event.x} {event.y}")
    
window = Tk()
frame = Frame(window,
             width=500,
             height=500,
             bg="yellow")
# 마우스 모션 이벤트 처리 함수를 바인딩 하고 있다.
frame.bind("<Motion>",motion)
frame.pack()

window.mainloop()
```

```python
# 이벤트가 발생하면 메시지 박스가 보여지는 예제

from tkinter import *
from tkinter import messagebox

# 이벤트 처리함수 정의
# 사용자가 입력한 내용을 가져와서 messagebox를 띄우는 함수
def click(event):
    test = entry.get()
    messagebox.showinfo("name",test)
    
window = Tk()
entry = Entry(window)
entry.pack(side=LEFT,padx=5)
button = Button(window,text="확인")
button.pack(side=LEFT)
# 버튼을 바인딩 처리
button.bind("<Button-1>",click)

window.mainloop()
```

```python
# 캔버스에 마우스 이벤트를 이용하여 선그리기 실습

from tkinter import *
window = Tk()
canvas = Canvas(window,
               width=500,
               height=500,
               bg = "white")


def draw(event):
    global x,y
    canvas.create_line(x,y,event.x,event.y)
    x,y = event.x,event.y
    
def down(event):
    global x,y
    x,y = event.x, event.y
    
    
def up(event):
    global x,y
    if (x,y) == (event.x,event.y):
        canvas.create_line(x,y,x+1,y+1)
        
canvas.bind("<B1-Motion>",draw)
canvas.bind("<Button-1>",down)
canvas.bind("<ButtonRelease-1>",up)
canvas.pack()

window.mainloop()
```
