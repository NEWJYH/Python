```python
# 기능이 없는 메뉴를 생성하는 실습

from tkinter import *

# 윈도우 생성
window = Tk()

# 메인메뉴생성
# mainMenu 변수에 Menu 생성 . 소유자 window
mainMenu = Menu(window)

# 윈도우에 메뉴는 mainMenu이라고 지정 하는 것
window.config(menu=mainMenu) 

# tearoff = False, or tearoff=0 <- 메뉴 제일 위 ---점선 제거
fileMenu = Menu(mainMenu, tearoff=False)
# 다른 메뉴가 확장이 되어야 되기 떄문에 add_cascade() 사용하여
# 메뉴의 이름을 "파일"로 설정하였고
# menu 속성값을 fileMenu로 지정하였다.
mainMenu.add_cascade(label="파일",menu=fileMenu)

# 하위 메뉴를 만들어 보기
fileMenu.add_command(label="열기")
# 구분선
fileMenu.add_separator()
fileMenu.add_command(label="종료")

window.mainloop()
```

```python
# 기능을 추가하는 메뉴를 생성하는 실습
from tkinter import *

from tkinter import messagebox

# 함수 정의
def func_open():
    messagebox.showinfo("메뉴선택","열기 메뉴를 선택함")
    
def func_exit():
    # 윈도우 종료
    window.quit()
    # 윈도우를 메모리에서 제거
    window.destroy()


# 윈도우 생성
window = Tk()

# 메인메뉴생성
# mainMenu 변수에 Menu 생성 . 소유자 window
mainMenu = Menu(window)

# 윈도우에 메뉴는 mainMenu이라고 지정 하는 것
window.config(menu=mainMenu) 

# tearoff = False, or tearoff=0 <- 메뉴 제일 위 ---점선 제거
fileMenu = Menu(mainMenu, tearoff=False)
# 다른 메뉴가 확장이 되어야 되기 떄문에 add_cascade() 사용하여
# 메뉴의 이름을 "파일"로 설정하였고
# menu 속성값을 fileMenu로 지정하였다.
mainMenu.add_cascade(label="파일",menu=fileMenu)

# 하위 메뉴를 만들어 보기
# 열기 버튼을 클릭하면 func_open 함수가 호출
fileMenu.add_command(label="열기",command=func_open)
# 구분선
fileMenu.add_separator()
# 종료 버튼을 클릭하면 func_exit 함수가 호출
fileMenu.add_command(label="종료",command=func_exit)

window.mainloop()
```

```python
# 대화상자의 생성과 사용에 대한 실습

from tkinter import *
# 사용자로부터 입력을 받는 대화상자를 사용하기 위한 모듈 
from tkinter.simpledialog import *

window = Tk()
window.geometry("400x100")

lbl1 = Label(window,text="입력된 값") # 기본 텍스트 값을 지정함.
lbl1.pack()

# simpledialog 모듈에는 사용자로부터 숫자(askinteger)
# 문자열(askstring), 실수(askfloat)를 입력받을 수 있다.
value = askinteger("주사위",
                   "주사위 숫자(1~6)를 입력하세요",
                   minvalue=1,
                   maxvalue=6)
# 사용자로부터 입력 받은 값을 lbl1에 표식이 된다.
lbl1.configure(text=str(value))


window.mainloop()
```
