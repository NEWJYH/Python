```python
from collections import * 
Person = namedtuple("Person","name age gender")
p1 = Person._make(["신은혁",12,"남성"])
p2 = Person._make(["김연아",31,"여성"])

# 클래스들에서 _fields 라고 출력할수 있음
# 생성되어진 Person이라는 namedtuple의  _fields로써
# Person에 있는 필드명을 tuple()형식으로 리턴을 해준다.
print(p1._fields)
```

```python
#getattr : 인스턴스의 값들을 가져와줌
print(getattr(p1,"name"))
# getattr()는 필드명으로 그 값을 출력할때 사용한다.
```

```python
# **(double-star-operator)은 namedtuple() 딕셔너리 자료구조를
# namedtuple로 변환하여 반환한다.
dic = {"name":"신은혁","age":5,"gender":"남성"}
print(dic)
print(type(dic))
p2 = Person(**dic)
print(p2)
print(type(p2))
```

```python
# 텍스트 마이닝 기법

def file_read(text):
    f_name = open("15.law.txt",mode="r",encoding="utf-8")
    for line in f_name:
        words = line.strip().split()
        for word in words:
            if len(word)>=2:
                text.append(word)
    print(text)


if __name__=="__main__":
    from collections import *
    text = []
    file_read(text)
    word_count = defaultdict(lambda: 0) # 값을 0으로 초기화
    
    # 텍스트에 담긴 단어들의 빈도수를 증가시키는 부분
    for word in text:
        word_count[word] += 1
        
    for i, v in OrderedDict(sorted(word_count.items(),key = lambda t:t[1],reverse = True)).items():
        if v >= 2:
            print(i,v)            
```
