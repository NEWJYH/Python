```python
# 캔버스를 생성하여 직선, 원, 사각형, 다각형 등 실습하기

# 캔버스 위젯은 파이썬에서는 그래픽 기능을 제공하는 것이다.

from tkinter import *

window = Tk()
# window.geometry("700x700")
canvas = Canvas(window, width=500, height=500, bg="white")
canvas.pack()

# 선 그리기
# 선에서 fill은 색상이고, width는 선의 굵기가 된다.
canvas.create_line(0,0,
                   500,500,
                   fill="BLUE")

canvas.create_line(0,500, # 0,500좌표에서 시작
                   500,0, # 500,0 좌표 까지
                   fill="RED", # 선의 색깔
                   width=5) # 선의 굵기


window.mainloop()
```

```python
from tkinter import *

window = Tk()
canvas = Canvas(window, width=500, height=500, bg="white")
canvas.pack()

# 사각형 그리기
# 사각형에서 fill은 채우기 색상이고,

canvas.create_rectangle(50,50, # 사각형의 왼쪽 상단 좌표
                       200,200, # 사각형의 오른쪽 하단 좌표
                       fill="yellow") # 색채우기 

# width는 바깥선의 굵기가 된다.
# outline 값은 바깥선의 색상이 된다.
canvas.create_rectangle(200,200, # 사각형의 왼쪽 상단 좌표
                       300,400, # 사각형의 오른쪽 하단 좌표
                       fill="red", # 색채우기 
                       outline="blue", # 외각 선 파란색으로
                       width=10) #((와각)선 두께)

window.mainloop()
```

```python
# Canvas 의 메서드
# coords() : 좌표를 변경하는 메서드
# itemconfig() : 색상을 변경하는 메서드
# delete() : 선, 도형 등을 삭제하는 메서드

from tkinter import *

window = Tk()

canvas = Canvas(window, width=500, height=500, bg="white")
canvas.pack()


# 선 그리기
line1 = canvas.create_line(0,0,500,500,fill="BLUE")
line2 = canvas.create_line(0,500, 500,0, fill="RED", width=5)

# line1의 좌표를 변경한다.
canvas.coords(line1, 250, 250, 500, 500)
# line2의 색상을 변경한다.
canvas.itemconfig(line2,fill="orange")


# 사각형 그리기
rect1 = canvas.create_rectangle(50,50,200,200,fill="yellow") 
rect2 = canvas.create_rectangle(200,200,300,400,fill="red", outline="blue",width=10) 

# rect1을 삭제해보기
canvas.delete(rect1)

# 캔버스에 그려진, 선, 도형들을 다 삭제하기
canvas.delete(ALL)

window.mainloop()
```

```python
from tkinter import *

window = Tk()
canvas = Canvas(window, width=500, height=500, bg="white")
canvas.pack()

canvas.create_rectangle(10,10,300,300)

# 사각형의 내접한 타원 그리기
canvas.create_oval(10,10, # 사각형의 왼쪽 상단 좌표 
                  300,300, # 사각형의 오른쪽 하단 좌표
                  fill="BLUE", # 채우기 색상
                  outline="orange", # 바깥선
                  width=10) # 바깥선의 두꼐

# 사각형에 내접한 원 그리기(extent 중심각도는 360도는 안된다.)
canvas.create_arc(10,10, # 사각형의 왼쪽 상단 좌표
                  300,300, # 사각형의 오른쪽 하단 좌표
                  extent=90, # 내접한 원의 중심각도
                  fill="orange", # 채우기 색상 오렌지
                  outline="red", # 바깥선(외곽선) 레드
                  width=5) # 바깥선의 두께


window.mainloop()
```

```python
# 다각형 그리기
from tkinter import *

window = Tk()
canvas = Canvas(window, width=500, height=500, bg="white")
canvas.pack()

canvas.create_rectangle(10,10,300,300)

# 다각형 그리기
canvas.create_polygon(10,10 ,100,100, 150,70, fill="blue")

window.mainloop()
```

```python
# 글자 나타내기
from tkinter import *

window = Tk()
canvas = Canvas(window, width=500, height=500, bg="white")
canvas.pack()

canvas.create_text(100,350,text="안녕하세요",font=("고딕",15))

canvas.create_text(200,450,text="안녕하세요",font=("고딕 15 bold"))

window.mainloop()
```
