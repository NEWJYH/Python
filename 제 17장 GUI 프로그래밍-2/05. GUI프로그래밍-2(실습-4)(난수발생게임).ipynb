```python
# 숫자 추측게임 만들기
from tkinter import *
import random

answer = random.randint(1,100) # 1부터 100까지 모두 
print(f"난수 발생 : {answer}")

# 함수 구현부
# event <- 앤터키 바인딩을 위함
def guessing(event):
    # 사용자가 입력한 값을 엔트리에서 가져오고 있다.
    guess = int(guessField.get())
    if guess > answer :
        msg = "높음!"
    elif guess < answer:
        msg = "낮음!"
    else:
        msg = "정답!"
        reset()
    # 텍스트를 라벨의 값으로 설정함
    resultLabel["text"] = msg
    # Entry에 입력되어 있는 값을 지운다.
    guessField.delete(0,END)
    
def reset():
    global answer
    answer = random.randint(1,100)
    print(f"난수 발생 : {answer}")
    resultLabel["text"] = "초기화"



# 윈도우 생성
window = Tk()

# guessing 함수에 엔터키 바인딩
window.bind('<Return>', guessing)

# 루트 윈도우의 배경색을 설정
window.configure(bg="white")

# 윈도우 타이틀 설정
window.title("숫자 추측 게임") 

# 윈도우 크기 설정
window.geometry("500x100")

# 라벨 생성
titleLabel = Label(window,
                  text = "숫자 게임에 오신것을 환영 합니다.")
titleLabel.pack()


# 추측 값을 입력하는 엔트리를 만듬
# 엔트리(입력필드 설정 한줄단위) 생성
guessField =Entry(window)
# 엔트리 왼쪽 정렬
guessField.pack(side=LEFT,padx=15)

# 버튼을 만들고 이벤트 처리 등록
tryButton = Button(window,
                  text="시도",
                  fg = "green",
                  bg = "white",
                  command = guessing)

resetButton = Button(window,
                    text="초기화",
                    fg="red",
                    bg="white",
                    command = reset)

tryButton.pack(side=LEFT, padx=5)
resetButton.pack(side=LEFT, padx=5)

# 결과 라벨을하나 만듬
resultLabel = Label(window,
                   text = "1부터 100사이의 숫자를 입력하세요",
                   bg="white")
resultLabel.pack(side=LEFT)


window.mainloop()
```
