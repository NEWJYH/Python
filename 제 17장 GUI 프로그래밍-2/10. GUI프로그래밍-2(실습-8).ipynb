```python
# Canvas_shape

from tkinter import *
import random

class Canvas_shape:
    def __init__(self):
        self.color = ["red","orange","black","yellow","violet","green"]
        self.window = Tk()
        self.window.title("여러가지 도형 그리기")
        # 윈도우 창에 캔버스를 만들고 배치함
        self.canvas = Canvas(self.window,
                            width=500,
                            height=500,
                            bg="white")
        self.canvas.pack()
        # 여러가지 버튼을 배치하는 frame 컨테이너를 생성하고 윈도우에 배치하기
        self.frame = Frame(self.window,
                          width = 600,
                          height = 100,
                          bg ="black")
        self.frame.pack()
        # 도형을 만들어주는 여러가지 버튼을 생성하여 frame에 배치하기
        # 이벤트 처리 연결
        self.btnRectangle = Button(self.frame,
                                  text="사각형",
                                  command = self.displayRect)
        
        self.btnOval = Button(self.frame,
                             text="타원",
                             command=self.displayOval)
        
        self.btnArc = Button(self.frame,
                            text="호",
                            command=self.displayArc)
        
        self.btnPolygon = Button(self.frame,
                            text="다각형",
                            command=self.displayPolygon)
        
        self.btnLine = Button(self.frame,
                             text="선",
                             command=self.displayLine)
        
        self.btnString = Button(self.frame,
                               text="문자열",
                               command=self.displayString)
        
        self.btnClear = Button(self.frame,
                              text="전체삭제",
                              command=self.displayClear)
        
        # 위에서 만든 버튼을 격자형으로 배치한다.
        self.btnRectangle.grid(row=1,column=1)
        self.btnOval.grid(row=1,column=2)
        self.btnArc.grid(row=1,column=3)
        self.btnPolygon.grid(row=1,column=4)
        self.btnLine.grid(row=1,column=5)
        self.btnString.grid(row=1,column=6)
        self.btnClear.grid(row=1,column=7)
        self.window.mainloop()
    
    # 사각형 출력 메서드
    def displayRect(self):
        # tags 는 위젯의 특정 부분에 이름을 추기하는 속성이다.
        # 식별자가 되기도 하고
        # 그룹화 시킬때도 사용할수가 있다.
        x1 = self.randomint()
        y1 = self.randomint()
        x2 = self.randomint()
        y2 = self.randomint()
        self.canvas.create_rectangle(x1,y1,
                                     x2,y2,
                                     tags="rect")
        
    # 타원 출력 메서드
    def displayOval(self):
        x1 = self.randomint()
        y1 = self.randomint()
        x2 = self.randomint()
        y2 = self.randomint()
        self.canvas.create_oval(x1,y1,
                                x2,y2,
                                fill = self.randomcolor(),
                                tags="oval")
    # 호 그리기
    def displayArc(self):
        x1 = self.randomint()
        y1 = self.randomint()
        x2 = self.randomint()
        y2 = self.randomint()
        self.canvas.create_arc(x1,y1,
                               x2,y2,
                               extent= self.randomExtent(),
                               fill = self.randomcolor(),
                               tags="arc")
        
    # 다각형그리기
    def displayPolygon(self):
        x1 = self.randomint()
        y1 = self.randomint()
        x2 = self.randomint()
        y2 = self.randomint()
        x3 = self.randomint()
        y3 = self.randomint()
        x4 = self.randomint()
        y4 = self.randomint()
        self.canvas.create_polygon(x1,y1,
                                   x2,y2,
                                   x3,y3,
                                   x4,y4,
                                   fill = self.randomcolor(),
                                   tags="polygon")
        
    def displayLine(self):
        x1 = self.randomint()
        y1 = self.randomint()
        x2 = self.randomint()
        y2 = self.randomint()
        # activefill 속성은 라인에 마우스를 올렸을때 색상을 바꾸게 해주는 것이다.
        # arrow 속성은 
        # first, end, last(시작,끝, 양끝지점) 화살표 속성으로 나타낸다.
        self.canvas.create_line(x1,y1,
                                x2,y2,
                                fill = self.randomcolor(),
                                width = random.randrange(10),
                                arrow="last",
                                activefill = self.randomcolor(),
                                tags="line")
    
    def displayString(self):
        x1 = self.randomint()
        y1 = self.randomint()
        self.canvas.create_text(x1,y1,
                               fill = self.randomcolor(),
                               text = self.randomcolor(),
                               font=("Times 10 bold underline"),
                               tags="string")
        
    def displayClear(self):
        self.canvas.delete(ALL)
        
    
    def randomExtent(self):
        return random.randrange(0,360)
    
    def randomint(self):
        return random.randrange(500)
    
    def randomcolor(self):
        return random.choice(self.color)
    
    
myCanvas = Canvas_shape()
```
