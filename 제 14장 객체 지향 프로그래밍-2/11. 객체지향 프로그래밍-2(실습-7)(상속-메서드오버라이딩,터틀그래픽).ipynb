```python
import turtle
import random
# 조상 클래스 Shape 정의
class Shape:
    def __init__(self):
        # 터틀 인스턴스 생성
        self.myTurtle = turtle.Turtle("turtle")
        self.cx = 0
        self.cy = 0
        
    # 펜 색상과 두께를 무작위로 생성
    def setPen(self):
        r = random.random() # 0.0<= 값 <1.0
        g = random.random()
        b = random.random()
        print(f"r : {r}, g : {g}, b : {b}")
        self.myTurtle.pencolor((r,g,b)) # 펜 색상 지정하기
        pSize = random.randrange(1,20) # 펜의 굵기를 임의로 설정위함
        self.myTurtle.pensize(pSize) # 펜의 굵기를 지정
    
    # 도형그리기
    # Shape 클래스를 상속받는 클래스 들은
    # 필요에 의해서 drawShape()를 오버라이딩 하여 사용할수 있도록
    # 선언부 선언과 아무 내용이 없는 구현부를 만들어 두었음
    def drawShape(self): 
        pass
        
```

```python
# from Shape import *

# 자손 클래스 Rectangle 정의
class Rectangle(Shape):
    def __init__(self,x,y):
        super().__init__()
        self.cx = x
        self.cy = y
        # 사각형의 폭과 높이를 임의의 수로 결정
        self.width = random.randrange(20,100)
        self.height = random.randrange(20,100)
        print(f"width : {self.width}")
        print(f"height : {self.height}") 
    
    # 조상 클래스 drawShape()를 오버라이딩
    def drawShape(self):
        # 사각형 그리기
        sx1,sy1 = 0,0 # 좌측 상단 좌표값
        sx2,sy2 = 0,0 # 우측 하단 좌표값
        
        sx1 = self.cx - self.width/2
        sy1 = self.cy + self.height/2
        sx2 = self.cx + self.width/2
        sy2 = self.cy - self.height/2
        print(f"sx1 : {sx1}, sy1 : {sy1}")
        print(f"sx2 : {sx2}, sy2 : {sy2}")
        
        # 팬의 색상과 두께를 얻기 위해서 조상 클래스의 메서드 호출
        self.setPen() 
        self.myTurtle.penup() # 펜을 드는 메서드
        self.myTurtle.goto(sx1,sy1) # 펜을 좌측 상단으로 이동
        self.myTurtle.pendown()# 펜을 내리는 메서드
        self.myTurtle.goto(sx2,sy1) # 펜을 우측 상단으로 이동
        self.myTurtle.goto(sx2,sy2) # 펜을 우측 하단으로 이동
        self.myTurtle.goto(sx1,sy2) # 펜을 좌측 하단으로 이동
        self.myTurtle.goto(sx1,sy1) # 펜을 좌측 상단으로 이동        
```

```python
# 함수 선언 부분
def screenLeftClick(x, y):
    rect = Rectangle(x, y)
    rect.drawShape()
    
# 메인 코드 부분
turtle.title("객체 지향 사각형 그리기")
# 화면을 클릭시 매개변수로 들어온 좌표(x,y)를 
# 매개변수로 들어온 함수에 인자값으로 실행, 
# 아래코드는 터틀 그래픽 판에서 
# 마우스 왼쪽 버튼이 클릭이 되는 것을
# 감지하는 리스너 메서드이다.
# 1은 왼쪽 버튼, 2는 휠, 3은 우측 버튼
turtle.onscreenclick(screenLeftClick,1)
turtle.done() # 터틀창 자동 닫기 방지함
```
