```python
# 구의 부피를 계산하는 함수 sphereVolume()를 작성하여 보자.
# 그리고 반지름을 사용자로부터 입력을 받고 구의 부피를 구하는 함수를 호출해서
# 테스트하라.

# 구의 부피 4/3*파이*(r**2)

def sphereVolume(radius):
    import math
    return 4/3 * math.pi * math.pow(radius, 3)
    
radius = float(input("구의 반지름을 입력하세요 : \n"))
answer = sphereVolume(radius)

print(f"반지름 = {radius} 구의 부피 = {round(answer,2)}")
```

```python
# 일회용 패스워드 생성기를 이용하여서 3개의 패스워드를 생성하여 출력하는
# 프로그램을 작성해보자.
# 패스워드의 길이는 6개 자리로 한정한다.
# 난수가 발생되는 범위 값을 지정을 아래와 같이 한다.
# 난수 발생 문자열 : '0123456789'

def disposable_password():
    import random
    password =""
    for i in range(6):
        password += str(random.randrange(0,10))
    return password
    
your_password = disposable_password()
print("본인 인증을 위해서 발송한 숫자를 입력하세요 : ", your_password)

def get_Password():
    import random
    num_str ="0123456789"
    password = ""
    
    for i in range(6):
        index = random.randrange(len(num_str))
        password += num_str[index]
    return password
    
print("본인 인증을 위해서 발송한 숫자를 입력하세요 : ", get_Password())
```

```python
# 사용자로부터 10진수를 입력받아서 
# 2진수로 변환하여 반환하는 프로그램을 작성하시오
# 함수 decTobin(num) 

def decTobin(num):
    answer = ""
    while True:
        if num % 2 == 0:
            answer = "0" + answer 
        else:
            answer = "1" + answer
        num  = num // 2
        
        if num == 0 :
            answer = "0" + answer 
            break
        if num == 1:
            answer = "1" + answer 
            break
            
    return answer

answer = decTobin(30)
print(answer)
```

```python
def dectobin(num):
    binary = ""
    
    while num != 0 :
        value = num % 2
        if value == 0:
            binary = "0" + binary
        else:
            binary = "1" + binary
        num = num // 2
        
    return binary
        
decNum = int(input("10진수를 입력하세요 : "))
print(f"10진수 {decNum} = 2진수 {dectobin(decNum)}")
print(f"파이썬 제공 2진수로 변환하는 내장 함수 : bin({decNum}) = {bin(decNum)}")

print("파이선 내장함수")
print("8진수")
print(oct(decNum))
print("16진수")
print(hex(decNum))
```

```python
# 10진수로 입력받은 숫자를 2진수 8진수 16진수로 리턴해주는 함수 구현

def my_convert_num(num):
    num_2 = bin(num)
    num_8 = oct(num)
    num_16 = hex(num)
    print(f"10진수 {num:>10} 2진수 {num_2:>10} 8진수 {num_8:>10} 16진수 {num_16:>10}")
    return num_2,num_8,num_16

for i in range(0,101):
    my_convert_num(i)
```
