```python
# 모듈에 대한 실습

# 모듈 : 파이썬에서는 모듈(module)은 변수나 함수, 그리고 클래스 들을 모아 놓은 파일
# 모듈은 반드시 확장자가 .py가 되어야 한다.

# 모듈을 사용하는 이유가 유지보수를 쉽게 하기 위해서
# 여러개의 파일로 분리를 할수가 있다.
# 모듈에 정의 및 구현해 놓은 함수를 따로 작성하지 않고
# 가져다가 여러 프로그램에서 사용하면 된다.
# 이것이 바로 코드의 재활용이다.
# 모듈을 사용하기 위해서는 import라는 키워드를 사용하여 모듈을 파일로 가져오면 된다.
```

```python
# 피보나치 수열 모듈 만들기
# 정수 형태로 피보나치 수열을 출력하는 함수
def fib(n : int):
    a,b = 0,1
    while b < n:
        print(b, end = " ")
        a, b = b, a+b
    print()
    
def fib_list(n : int):
    result = []
    a,b = 0,1
    while b < n:
        result.append(b)
        a,b = b, a+b
    return result
```

```python
# fibo 모듈을 가져오고 있음
import fibo 
fibo.fib(1000)
print(fibo.fib2(100))
```

```python
# 유용한 모듈 사용해보기
# copy모듈 : 객체를 복사할 떄 사용하는 모듈이다.

# 1. 얕은 복사(shallow copy) : 
#             객체의 참조값(주소)만 복사 되고 객체는 공유가 되버린다.
#             원본 객체에 영향을 끼친다. 완벽한 복사가 아니다.

# 2. 깊은 복사(deep copy):
#             독립된 객체를 복사해준다.
#             원본 객체에 영향을 미치지 않는다.

import copy
colors = ["red","blue","green"]
clone = copy.deepcopy(colors) # 깊은 복사가 발생한다.
clone[0] = "빨간색"
print(colors)
print(clone)
```

```python
import random
# random 모듈 사용해보기
# random 모듈은 난수를 발생할때 사용하는 모듈이다.

# randint() : 정수 범위의 난수를 발생하고자 할 때 사용한다.
#           : randint(a,b)이면 a부터 b까지 랜덤하게 출력
myset = set()
i = 0 
while len(myset) != 6:
    a = random.randint(1, 6)
    myset.add(a)
    i += 1
    print(f"{i}번 째 반복중")
print(myset)

# random() : 0.0 ~ 1.0미만의 난수를 발생시킨다.
# 난수의 범위를 정하고 싶다면 원하는 범위의 수를 곱하면 된다.
print(f"random() : {random.random()*100}")

# randrange() : randrange(start,stop,[step]) start부터 stop전까지
#               사이의 값을 랜덤하게 발생
myset = set()
i = 0 
while len(myset) != 100:
    a = random.randrange(1, 101)
    myset.add(a)
    i += 1
    print(f"{i}번 째 반복중")
print(myset)

# choice() : 주어진 시퀀스의 항목을 랜덤하게 가져온다.
list1 = ["신은혁","김연아","손연재","추신수"]
print(random.choice(list1))

# shuffle() : 리스트의 항목을 랜덤하게 섞어주는 함수이다.
list2 = [[x]for x in range(10)]
print(list2)
random.shuffle(list2)
print(list2)
```

```python
# time 모듈
import time
# time() : 1970년 1월 1일 0시 이후 지금까지 흐른 절대 시간을 초단위로 출력함
# 프로그램의 성능테스트 등에 자주 사용함.
print(f"{time.time()}")


def fib(n : int):
    a,b = 0,1
    while b < n:
        print(b, end = " ")
        a, b = b, a+b
    print()

start = time.time() # 시작값
fib(1000000)
end = time.time() # 종료값
exetime = end - start
print(start)
print(end)
print(f"{exetime:.9f}")

# asctime() : 현재 날짜 및 시간을 보여줌
#             현재 날짜와 시간을 문자열 형태로 표시해준다.
print(time.asctime())

# sleep() : 프로그램을 일시 정지 시키는 함수
for i in range(10,0,-1):
    print(i,end=" ")
    time.sleep(1)
```
