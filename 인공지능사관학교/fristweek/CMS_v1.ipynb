```python
# 고객관리시스템 v0.1
# 데이터 : 이름, 성별, 이메일, 출생연도
# 이름은 문자열; 성별은 M/F; 이메일은 문자열; 연도는 정수;

customer_list =[]

while True:
    menu = '''안녕하세요?
    입력은 I 를 누르세요.
    수정은 U 를 누르세요.
    삭제는 D 를 누르세요.
    조회는 L 을 누르세요.
    전체 조회는 E 를 누르세요.
    종료는 X 를 누르세요.'''
    print(menu)
    choice = input().upper()
    print(choice)

    
    if choice == 'I':
        print('이름을 입력하세요.')
        name = input()
        print('성별을 입력 하세요')
        gender = input()
        print('이메일을 입력 하세요')
        email = input()
        print('출생연도를 입력 하세요')
        year = int(input())
        # DICT
        info ={}
        info['name']= name
        info['gender']= gender
        info['email']= email
        info['year']= year
        customer_list.append(info) 
        print(customer_list)
    
    elif choice =='E':
        for dict in customer_list:
            print(dict)

    elif choice == 'U':
        b = input('조회 할 고객 성함을 적어주세요.')
        num = 0
        for dict in customer_list:
            if dict['name'] == b:
                print(dict)
                break
            num = num + 1
            
        c = input('무엇을 변경하시겠습니까? 1. 이메일 2. 성별 3. 출생연도')
        if c == '1':
            d = input('변경할 이메일을 작성해주세요')
            customer_list[num]['email'] = d
            print(customer_list)
        elif c == '2':
            d = input('변경할 성별은?')
            customer_list[num]['gender'] = d
            print(customer_list)
        elif c == '3':
            d = input('변경할 출생년도는?')
            customer_list[num]['year'] = d
            print(customer_list)
        else :
            print('잘못 입력하셨습니다')

    elif choice == 'D':
        b = input('삭제 할 고객 성함을 적어주세요.')
        num = 0
        for dict in customer_list:
            if dict['name'] == b:
                print(dict)
                break
            num = num + 1
        customer_list.pop(num)
    
    
    elif choice == 'L':
        b = input('조회 할 고객 성함을 적어주세요.')
        for dict in customer_list:
            if dict['name'] == b:
                print(dict)
                break
            
    
    elif choice == 'X':
        break
    
    else:
        print('잘못된 입력입니다!')



        
```
