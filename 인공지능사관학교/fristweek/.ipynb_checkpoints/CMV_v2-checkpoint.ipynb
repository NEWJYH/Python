```python
# 고객관리 시스템 v0.2

# 고객 정보를 담을 리스트
customer_list = []
# 메뉴 옵션
choice_list = "I U D L X".split()



# 함수 입력부
def name_Input():
    return input("이름을 입력하세요\n")

def gender_Input():
    gender = input("성별을 입력하세요 (남자 : M, 여자 : F)\n").upper()
    if gender == "M" or  gender == "F":
        return gender 
    else:
        print("잘못된 입력입니다.")
        return gender_Input()

def email_Input():
    email = input("email을 입력해주세요\n")
    if "@" in email:
        return email
    else:
        print("잘못된 입력입니다.")
        return email_Input()

def year_Input():
    year = input("출생연도를 입력해주세요 (정수)\n")
    try:
        year = int(year)
        return year
    except:
        print("잘못된 입력입니다.\n정수로 입력해주세요")
        return year_Input()

# 회원정보가 담긴 딕셔너리 생성
def dict_Input(name, gender, email, year):
    return {"name":name, "gender":gender, "email":email, "year": year}

# 조회 선택
def search_list(flag):
    name = input(f"{flag} 할 고객 성함을 입력해주세요.\n")
    idx = -1
    for i in range(0, len(customer_list)):
        if customer_list[i]["name"] == name:
            idx = i
            break
    if idx == -1:
        print("데이터가 존재하지 않습니다.")
    else :
        print(f'''
        이름 : {customer_list[idx]['name']}
        성별 : {customer_list[idx]['gender']}
        email : {customer_list[idx]['email']}
        출생연도 : {customer_list[idx]['year']}
            ''')
        if flag != "조회":
            return idx

# 수정 선택
def modify_list_eliment(idx):
    user_select = input("변경가능 목록\n 1. 성별\n 2.email\n 3.출생연도 ")
    if user_select == "1":
        customer_list[idx]["name"] = name_Input()
    elif user_select == "2":
        customer_list[idx]["email"] = email_Input()
    elif user_select == "3":
        customer_list[idx]["year"] = year_Input()
    else:
        print("잘못된 선택입니다.")
        return modify_list_eliment(idx)

# 삭제 선택
def delete_eliment(idx):
    if  idx == None or idx >= len_list():
        pass
    else:
        del customer_list[idx]
        

def len_list():
    return len(customer_list)
    
    
# main()    
print("고객관리 시스템 v0.2")    
# mainLoop
while True:
    
    menu = '''
    
    
    입력 I
    수정 U
    삭제 D
    조회 L
    종료 X
    
    
    '''
    # 메뉴 입력 
    choice = input(menu).upper()
    if choice not in choice_list:
        print("잘못된 입력입니다.")
        continue
    # 입력 선택
    if choice == "I":
        name = name_Input()
        gender = gender_Input()
        email = email_Input()
        year = year_Input()
        # Dictionary 형태로 리스트에 추가
        customer_list.append(dict_Input(name, gender, email, year))
    # 조회 선택
    elif choice == "L":
        search_list("조회")
    # 수정 선택
    elif choice == "U":
        modify_list_eliment(search_list("변경"))
    # 삭제 선택
    elif choice == "D":
        delete_eliment(search_list("삭제"))
    # 프로그램 종료
    elif choice == "X":
        break
```
