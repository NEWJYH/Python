1. SQL이란?
SQL이란 Structured Query Language의 약자로, 데이터베이스에 접근하고 조작하기 위한 표준 언어입니다.
SQL은 여러 개의 연결된 테이블로 이루어진 관계형 데이터베이스를 제어할 수 있게 해줍니다.
모든 테이블은 고유의 이름을 가지고 있으며, 각 테이블은 컬럼(Column)과 레코드(Record)로 구성되어 있습니다.



2. 테이블에서 데이터 검색하기
Point I
SELECT: 검색할 컬럼을 지정합니다.
FROM: 원하는 테이블을 지정합니다.

SELECT title, author 
--검색할 컬럼

FROM book; 
--테이블
Point II
검색할 데이터에 *을 입력하면 모든 데이터를 가져옵니다.

SELECT * 
FROM book; 
Point III
DISTINCT: 중복된 컬럼을 제거하고 보여줍니다.

SELECT DISTNCT title, author

FROM book;
명령어 뒤에 2개 이상의 컬럼을 적으면, 한 쪽 컬럼에 중복이 있어도 다른 쪽 컬럼의 값이 다르면 다르게 취급합니다.


3. 조건을 추가하여 검색하기
WHERE: 검색하고자 하는 데이터의 조건을 설정합니다.

SELECT * FROM book
WHERE title = '돈키호테'; --조건


4. 여러 개의 조건 추가하기
여러 개의 조건을 만족하는 데이터를 검색하기 위해 연산자를 사용합니다.

Point I


SELECT * FROM score
WHERE korean >= 90; --조건
score 테이블에서 국어 성적이 90 이상인 값 검색
Point II


SELECT * FROM score
WHERE korean >= 90 OR math > 80; --조건
score 테이블에서 국어 성적이 90 이상이거나 수학 성적이 80 초과인 값 검색
Point III


SELECT * FROM score
WHERE math BETWEEN 80 AND 90; --조건
score 테이블에서 수학 성적이 80 이상 90 이하인 값 검색
