01. 서브쿼리

서브쿼리 정의

하나의 쿼리 안에 포함된 또 하나의 쿼리

메인 쿼리가 서브쿼리를 포함하는 종속적인 관계

서브쿼리의 특징

- 알려지지 않은 기준을 이용한 검색에 유용

- 메인 쿼리가 실행되기 이전에 한 번만 실행

- 한 문장에서 여러 번 사용 가능

기존의 검색 방법

Example
SELECT * FROM employees
WHERE 급여 > 2500;


사원 elice의 급여를 알고 있는 상태에서 더 높은 급여를 받는 사원을 조회

서브쿼리를 사용한 예시

Example
SELECT * FROM employees                         <- 메인쿼리
WHERE 급여 >                                     <- 메인쿼리
(SELECT 급여 FROM employees WHERE 이름='elice');  <- 서브쿼리

사원 elice의 급여를 알지 못해도 검색 가능

서브쿼리 사용시 주의사항

1. 서브쿼리는 괄호와 함께 사용되어야 한다.

2. 서브쿼리 안에서 ORDER BY 절은 사용할 수 없다.

3. 서브쿼리는 연산자의 오른쪽에 사용되어야 한다.

4. 서브쿼리는 오로지 SELECT문으로만 작성 할 수 있다.

서브쿼리 기본 문법

Example
SELECT * FROM employee
WHERE 급여 >  ← 연산자 오른쪽에 서브쿼리
(SELECT 급여 FROM employee WHERE 이름='elice');

02. 반환에 따른 분류

단일 행 서브쿼리 정의

결과가 한 행만 나오는 서브쿼리
서브쿼리가 결과를 1개의 값만 반환하고,
이결과를 메인쿼리로 전달하는 쿼리

단일 행 서브쿼리 기본 문법
Example
SELECT * FROM employee
WHERE 급여 >
(SELECT 급여 FROM employee WHERE 사원번호 = 1) ;

사원번호는 기본적으로 1개만 있으므로 한 개의 행만 반환함 = 단일행

단일 행 서브쿼리 연산자

기호           뜻
 =             같다
 <>            같지 않다
 >             크다
 >=            크거나 같다
 <             작다
 <=            작거나 같다

다중 행 서브쿼리 정의

결과가 한 행만 나오는 단일 행 서브쿼리와는 다르게
서브쿼리가 결과를 2개 이상 반환하고,
이 결과를 메인쿼리로 전달하는 쿼리

다중 행 서브쿼리 기본 문법

Example
SELECT * FROM employee
WHERE 급여 IN (
SELECT max(급여) FROM employee GROUP BY 부서번호) ; 


- IN : 다중행 연산자

다중 행 연산자

기호      뜻
IN       하나라도 만족하면 반환
ANY      하나라도 만족하면 반환 비교 연산 가능
ALL      모두 만족하면 반환 비교 연산 가능


- < any 최대값    > all 최대값
- > any 최소값    < all 최소값  

다중 행 연산자 사용 예시

1 in (1,2,3,4)
    - 참
10 < any (1,2,3,4) : 최대값 -> 4
    - 10 < 4 거짓
99 >= all (99, 100, 101) : 최대값 (포함) -> 101
    - 99 >= 101  거짓

다중 행 서브쿼리 1
다중 행 서브쿼리(Multiple-Row Subquery) 는 서브쿼리의 조회 결과가 두개 이상의 row가 나오는 서브쿼리를 뜻합니다.

다중 행 비교 연산자

연산자	연산자 뜻
IN	하나라도 만족하면 참
ANY	내부적으로 모두 or 연산을 함
ALL	내부적으로 모두 and 연산을 함

지시사항
emp테이블에서 'MANAGER' 업무를 가진 사원 중 제일 높은 급여를 받는 사원보다 높은 급여를 받는 사원의 모든 컬럼을 조회하는 쿼리를 작성 해주세요.


```python
'''
SELECT * FROM emp;
-- MANAGER 업무를 가진 사원 중 제일 높은 급여를 받는 사원보다 높은 급여를 받는 사원을 조회하는 쿼리를 작성해주세요.

SELECT * 
FROM emp
WHERE sal > 
(SELECT MAX(sal)
FROM emp
WHERE job = 'MANAGER');

-- 서브쿼리 : 메니저 업무를 가진 사원중, 제일 높은 급여를 받는 사원

-- 메인쿼리 : 그 사원보다 높은 급여를 받는 사원

-- 메인쿼리 작성
select * 
from emp
where sal > all
-- 서브쿼리 작성
(select max(sal)
from emp
where job ='MANAGER') ;

'''

'''
+--------+--------+-----------+------------+------+--------+
| empnum | name   | job       | birthdate  | sal  | deptno |
+--------+--------+-----------+------------+------+--------+
|      1 | SMITH  | CLERK     | 1990-12-17 |  900 |     10 |
|      2 | ALLEN  | SALESMAN  | 1982-02-20 | 1600 |     20 |
|      3 | WARD   | SALESMAN  | 1983-02-22 | 1250 |     30 |
|      4 | JONES  | MANAGER   | 1981-04-02 | 2975 |     40 |
|      5 | MARTIN | SALESMAN  | 1981-09-28 | 1250 |     10 |
|      6 | BLAKE  | MANAGER   | 1981-05-01 | 2850 |     20 |
|      7 | CLARK  | MANAGER   | 1981-06-09 | 2450 |     30 |
|      8 | SCOTT  | ANALYST   | 1987-07-13 | 2000 |     40 |
|      9 | KING   | PRESIDENT | 1982-11-17 | 5000 |     10 |
|     10 | TURNER | SALESMAN  | 1983-09-08 | 1900 |     20 |
|     11 | ADAMS  | CLERK     | 1987-07-13 | 1100 |     30 |
|     12 | JAMES  | CLERK     | 1990-12-03 |  950 |     40 |
|     13 | FORD   | ANALYST   | 1981-12-03 | 3000 |     10 |
|     14 | MILLER | CLERK     | 1982-01-23 | 1300 |     20 |
+--------+--------+-----------+------------+------+--------+
+--------+------+-----------+------------+------+--------+
| empnum | name | job       | birthdate  | sal  | deptno |
+--------+------+-----------+------------+------+--------+
|      9 | KING | PRESIDENT | 1982-11-17 | 5000 |     10 |
|     13 | FORD | ANALYST   | 1981-12-03 | 3000 |     10 |
+--------+------+-----------+------------+------+--------+
+--------+------+-----------+------------+------+--------+
| empnum | name | job       | birthdate  | sal  | deptno |
+--------+------+-----------+------------+------+--------+
|      9 | KING | PRESIDENT | 1982-11-17 | 5000 |     10 |
|     13 | FORD | ANALYST   | 1981-12-03 | 3000 |     10 |
+--------+------+-----------+------------+------+--------+
코드 실행이 완료되었습니다.
'''
```
