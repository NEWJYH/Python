01. 관계,논리 연산자

관계 연산자

두 값을 비교하여 참 또는 거짓 값을 가진다.

Example
4 > 3; // true
7 < 5; // false
4 >= 4; // true
9 <= 8; // false
3 == 3; // true
3 != 4; // true
4 != 4; // false

논리 연산자

AND 연산자

Example
true && true // true
true && false // false
false && true // false
false && false // false

논리 연산자

OR 연산자

Example
true || true // true
true || false // true
false || true // true
false || false // false

논리 연산자

관계 연산자와 논리 연산자를 함께 사용해봅시다!

Example
(2 > 1) && (1 > 0) // true
(2 > 1) && (0 > 1) // false
(0 > 1) && (1 > 2) // false

논리 연산자

부정 연산자

Example
!true // false
!false // true
(0 > 1) // false
!(0 > 1) // true

명제 만들기 (1)
Java의 관계 연산자는 두 자료를 비교해 true, false중 하나를 반환합니다.

boolean b1 =   5 == 5;
boolean b2 =   3 == 5;

5와 5는 같은 수이므로 b1의 값은 true가 됩니다.

3과 5는 다른 수이기 때문에 b2의 값은 false가 됩니다.

수를 비교해보고 true와 false 중 원하는 결과가 나오도록 비교연산자를 채워 넣는 실습을 해봅시다.

관계 연산자	연산 특성
A < B	A가 B보다 작으면 true
A <= B	A가 B보다 작거나 같으면 true
A > B	A가 B보다 크면 true
A >= B	A가 B보다 크거나 같으면 true
A == B	A와 B가 같으면 true
A != B	A와 B가 같지 않으면 true

지시사항
**b1의 값이 true**가 될 수 있도록 ?을 지우고 맞는 관계 연산자를 넣으세요.

**b2의 값이 false**가 될 수 있도록 ?을 지우고 맞는 관계 연산자를 넣으세요.

**b3의 값이 true**가 될 수 있도록 ?을 지우고 맞는 관계 연산자를 넣으세요.

출력 예시

true, false, true

```python
'''
public class Main {
	public static void main(String args[]) {
        // 지시사항 1번을 참고하여 코드를 작성하세요.
        boolean b1 = 4 < 6;
        
        // 지시사항 2번을 참고하여 코드를 작성하세요.
        boolean b2 = 5 != 5;
        
        // 지시사항 3번을 참고하여 코드를 작성하세요.
        boolean b3 = 6 > 4;
        
        System.out.println(b1 + ", " + b2 + ", " + b3);
	}
}

'''

#
'''
true, false, true
'''
```

성적 계산하기
if문을 사용하면 조건에 따라 다른 결과를 출력하는데 사용되며 조건식의 결과가 참(true)이거나 거짓(false)일 때 실행문이 결정됩니다.

조건문의 구조는 다음과 같습니다.

if (조건식 1) {     
    실행문1;
else if (조건식 2) {     
    실행문2; 
} 
else {     
    실행문3; 
} 
Copy
위 코드를 살펴보면 if의 경우 만약 조건식 1이 참(true)이면 실행문 1을 실행합니다.

else if의 경우 if문이 거짓(false)이고 조건식2가 참(true)이면 실행문 2를 실행합니다.

else의 경우 위 두 조건식이 모두 거짓(false)이면 실행문3을 실행합니다.

if - else if - else문의 경우 조건에 따라 하나의 실행문만 실행됩니다.

위 코드를 참고하여 성적에 따라 학점을 출력해주는 프로그램을 작성해 봅시다.


지시사항
if를 사용하여 변수 grade가 90 이상이면 “학점은 A입니다.”를 출력하는 코드를 작성하세요.

else if를 사용하여 변수 grade가 80 이상이면 “학점은 B입니다.”를 출력하는 코드를 작성하세요.

else if를 사용하여 변수 grade가 70 이상이면 “학점은 C입니다.”를 출력하는 코드를 작성하세요.

else if를 사용하여 변수 grade가 60 이상이면 “학점은 D입니다.”를 출력하는 코드를 작성하세요.

else를 사용하여 변수 grade가 위 조건을 모두 만족하지 않는 경우 “학점은 F입니다.”를 출력하는 코드를 작성하세요.

입력 예시1

90

출력 예시1

학점은 A입니다.

입력 예시2

50

출력 예시2

학점은 F입니다.

```python
'''
package elice;

import java.util.Scanner;

public class Main {
	public static void main(String args[]) {
        Scanner input = new Scanner(System.in);
        int grade = input.nextInt();
        // 지시사항 1번을 참고하여 코드를 작성하세요.
        if (grade >= 90){
            System.out.println("학점은 A입니다.");
        }
        
        // 지시사항 2번을 참고하여 코드를 작성하세요.
        else if (grade >= 80){
            System.out.println("학점은 B입니다.");
        }
        
        // 지시사항 3번을 참고하여 코드를 작성하세요.
        else if (grade >= 70){
            System.out.println("학점은 C입니다.");
        }
        // 지시사항 4번을 참고하여 코드를 작성하세요.
        else if (grade >= 60){
            System.out.println("학점은 D입니다.");
        }
        
        // 지시사항 5번을 참고하여 코드를 작성하세요.
        else{
            System.out.println("학점은 F입니다.");
        }
    input.close();
    }
}
'''
```

달리기 등 수
switch문의 경우 if문처럼 조건에 따라 값을 다르게 출력할 수 있지만 if문과는 다르게 변수가 어떤 값을 갖느냐에 따라 실행문이 결정됩니다.

switch문의 구조는 다음과 같습니다.

switch(변수){
    case 값1: 
        실행문1; 
        break;
    case 값2:  
        실행문2;
        break;
    case 값3:
        실행문3:
        break;
    default: 
        실행문4;
        break;
}
Copy
switch의 괄호() 안에 있는 변수의 값이 각 case의 값과 동일하다면 해당 실행문을 실행합니다.

default의 경우 case에 있지 않은 값이 변수에 들어갈 때 실행됩니다.

break를 각 실행문이 끝나고 사용해줘야 다른 case에 있는 실행문이 실행되지 않습니다.

위 코드를 참고하여 달리기 등 수에 따라 출력을 다르게 하는 프로그램을 작성해 봅시다.

변수 rank가 1일 경우 “금메달 입니다!”를 출력하는 case 문을 작성하세요.

변수 rank가 2일 경우 “은메달 입니다!”를 출력하는 case 문을 작성하세요.

변수 rank가 3일 경우 “동메달 입니다!”를 출력하는 case 문을 작성하세요.

위 조건을 모두 만족하지 않을 경우 “완주하였습니다!”를 출력하는 default 문을 작성하세요.

입력 예시1

1

출력 예시1

금메달 입니다!

입력 예시2

4

출력 예시2

Tips
break를 사용하지 않을 경우 다음 case가 연달아 실행됩니다.

```python
'''
package elice;

import java.util.Scanner;

public class Main {
	public static void main(String args[]) {
        Scanner input = new Scanner(System.in);
        System.out.println("달리기 등 수를 입력하세요.");
        int rank = input.nextInt();
       
        switch(rank){
        // 지시사항 1번을 참고하여 코드를 작성하세요.
        case 1 :
            System.out.println("금메달 입니다!");
            break;
    // 지시사항 2번을 참고하여 코드를 작성하세요.
        case 2 :
            System.out.println("은메달 입니다!");
            break;
        // 지시사항 3번을 참고하여 코드를 작성하세요.
        case 3 :
            System.out.println("동메달 입니다!");
            break;
        // 지시사항 4번을 참고하여 코드를 작성하세요.
        default:
            System.out.println("완주하였습니다!");
            break;
        }
    input.close();
    }
    
}
'''
```

장학금 받기
if문 안에 if문을 사용하여 조건 여러 개를 동시에 확인할 수 있습니다.

구조는 다음과 같습니다.

if (조건식 1) { 
    실행문 1;
    if(조건식 2) { 
        실행문 2;
    }
} 
else {     
    실행문 3;
} 

위 코드를 살펴보면 첫번째 if문의 조건식이 참(true)이면 실행문 1을 실행합니다.

그 다음 첫 번째 if문의 참(true)이면서 두 번째 if문이 참(true)이라면 실행문 2를 실행합니다.

만약 첫 번째 if문이 거짓(false)이라면 실행문 1,2가 둘다 실행이 되지 않고 else의 실행문 3이 실행됩니다.

위 구조를 참고하여 장학금 수여 여부를 판단해주는 프로그램을 작성해 봅시다.


지시사항
if를 사용하여 변수 grade가 70 이상이면 “합격입니다.”를 출력하는 코드를 작성하세요.

위 if문 안에 if를 사용하여 변수 grade가 90 이상이면 “장학금도 받을 수 있습니다.”를 출력하는 코드를 작성하세요.

else를 사용하여 변수 grade가 위 조건을 모두 만족하지 않는 경우 “불합격입니다.”를 출력하는 코드를 작성하세요.

입력 예시1

90

출력 예시1

합격입니다.
장학금도 받을 수 있습니다.

입력 예시2

80

출력 예시2

합격입니다.

```python
'''
package elice;

import java.util.Scanner;

public class Main {
	public static void main(String args[]) {
        Scanner input = new Scanner(System.in);
        int grade = input.nextInt();//성적을 입력받는 변수 grade
        // 지시사항 1번을 참고하여 코드를 작성하세요.
        if (grade >= 70){
            System.out.println("합격입니다.");
            // 지시사항 2번을 참고하여 코드를 작성하세요.
            if (grade >= 90){
                System.out.println("장학금도 받을 수 있습니다.");
            }
        }
        // 지시사항 3번을 참고하여 코드를 작성하세요.
        else{
            System.out.println("불합격입니다.");
        }
    input.close();
    }
}
'''
```

합계가 100이 넘어갈 때까지 더하기
continue 키워드는 반복문 내부에서 사용될 경우 그 키워드 이후에 있는 코드를 실행하지 않고 다음 조건식이나 증감식으로 이동합니다.

while(i < 10 ){
System.out.println("Elice");
continue;
i++;
}

위 코드의 경우 continue 키워드로 인해 i의 값이 증가하지 않아 “Elice” 문장이 무한정으로 출력됩니다.

break 키워드는 반복문을 종료할 때 사용합니다.

while(true){
System.out.println("Elice");
break;

위 코드의 경우 break 키워드로 인해 한번만 “Elice”를 출력하고 while문을 종료합니다.

위 키워드를 활용하여 1부터 숫자들을 차례대로 합산하되 합계가 100이 넘어갈 때까지 더해봅시다.

또 합계를 구할 때 3의 배수는 제외하고 더해봅시다.

예를 들어 1, 2, 4, 5 순서대로 더해나가다가 그 합계가 98일 때 한 번 더 더하고 100이 넘어가면 멈추는 것입니다.

이때의 합계가 정확히 몇인지 출력해 봅시다.


지시사항
continue키워드를 사용하여 변수 i를 3으로 나눈 나머지가 0인 경우 sum에 더하지 않는 if문을 작성하세요.

break 키워드를 사용하여 sum이 100을 넘어가면 종료하는 if문을 작성하세요.

Tips
나머지 연산자를 이용하여 3의 배수를 제외할 수 있습니다. 3의 배수는 3으로 나눈 나머지가 0이기 때문입니다.

```python
'''
package elice;

public class Main {
	public static void main(String args[]) {
        int i = 0;
        int sum = 0;
        
        while(true) {
            i++;
            // 지시사항 1번을 참고하여 코드를 작성하세요.
            if (i % 3 == 0){
                continue;
            }
            sum += i;
            // 지시사항 2번을 참고하여 코드를 작성하세요.
            if (sum >= 100){
                break;
            }
        }
        System.out.println(sum);
	}
}
'''
```
