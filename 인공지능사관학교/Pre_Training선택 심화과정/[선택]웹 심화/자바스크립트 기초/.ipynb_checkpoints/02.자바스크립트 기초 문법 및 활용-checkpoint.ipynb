문자열 +는 연결 연산자로 사용

Example
console.log("20" + "10"); // 2010
console.log("20" - "10"); // 10
console.log("20" * "10"); // 200
console.log("20" / "10"); // 2
console.log("20" % "10"); // 0

숫자 뿐 아니라 문자열도 산술 연산자 사용 가능

증감 연산자

var num = 10;

console.log(++num); // num + 1 후 num 출력
console.log(--num); // num - 1 후 num 출력

console.log(num++); // num 출력 후 num + 1
console.log(num--); // num 출력 후 num - 1

++ 1씩 증가           -- 1씩 감소

비교 연산자

Example
console.log(10 == 20);  // 값이 같다                     false
console.log(10 === 20); // 데이터 타입과 값이 같다        false
console.log(10 !== 20); // 값이 같지 않다                 true

console.log(10 > 20);    // false
console.log(10 >= 20);   // false
console.log(10 < 20);    // true
console.log(10 <= 20);   // true

== 와 === 차이

Example
console.log(10 == "10");  // true
console.log(10 === "10"); // false

Boolean 데이터 타입인
true 혹은 false 반환

논리 연산자

Example
// 앞 뒤 조건 모두 참인 경우에만 true 반환하는 AND 연산자
console.log(10 === 10 && 20 === 30); // false

// 둘 중 하나만 참이여도 true 반환하는 OR 연산자
console.log(10 === 10 || 20 === 30); // true

조건문 - else if 문
else if문은 여러 개의 조건문을 생성할 때 사용되는 조건문입니다.

var a = 3;

if ( a > b ) { 
 document.write("a는 b보다 크다.");
} else if ( b > c ) { 
 document.write("b는 c보다 크다.");
} else if ( a < c ) { 
 document.write("a는 c보다 작다.");
} else if ( b < c ) { 
 document.write("b는 c보다 작다.");
} else { 
 document.write("모든 조건을 만족하지 않는다.");
}
Copy
실습을 통해 else if 조건문을 직접 만들어 봅시다!


지시사항
var a = 20; var b = 40; var c = 60;을 입력하여 세 변수를 생성합니다.

조건문 형식 if ( ) { } else if ( ) { } else { }을 입력합니다.

만약 a가 b보다 크면, "a > b" 출력합니다.

만약 b가 c보다 크면, "b > c" 출력합니다.

만약 a가 c보다 작으면, "a < c" 출력합니다.

만약 b가 c보다 작으면, "b < c" 출력합니다.

위의 모든 조건에 부합하지 않으면, "모든 조건을 만족하지 않는다." 출력합니다.

Tips
else if문은 가장 첫 번째로 true가 나오는 조건의 명령을 수행합니다!

```python
'''
var a = 20;
var b = 40;
var c = 60;

if (a>b){ document.write("a > b");}
else if (b > c) {document.write("b > c");}
else if (a < c) {document.write("a < c");}
else if (b < c) {document.write("b < c");}
else { document.write("모든 조건을 만족하지 않는다.");}
'''

'''
a < c
'''
```

조건문 - 중첩 if 문
중첩 if문은 if문 안에 또 다른 if문을 삽입할 때 사용되는 조건문입니다.

var a = 3;
var b = 6;

if ( a !== b ) {

  if (a > b) { 
    console.log("a는 b보다 크다"); 
  } 
  else { 
    console.log("a는 b보다 작다"); 
  }

} else { 
  console.log("a와 b는 같다"); 
}
Copy
실습을 통해 중첩 if문을 직접 만들어 봅시다!

var a = 20; var b = 40;을 입력하여 두 변수를 생성합니다.

조건문 형식 if ( ) { if ( ) { } else { } } else { }를 입력합니다.

a와 b가 같지 않다면, 중첩된 if문 실행, 같다면 "a === b" 출력합니다.

중첩된 if문에서, a가 b보다 크면 "a > b", 그렇지 않으면 "a < b" 출력합니다.

```python
'''
var a = 20;
var b = 40;

if ( a !== b ) {

  if (a > b) { 
    document.write("a > b"); 
  } 
  else { 
    document.write("a < b"); 
  }

} else { 
  document.write("a === b"); 
}
'''

'''
a < b
'''
```

function isPrime(n) {
    var divisor = 2;                       // 2부터 나누기 시작
    while( n > divisor) {                  // n이 나누는 수보다 클 때까지
        if ( n % divisior === 0) {         // n과 나누는 수가 나누어 떨어지면
            return false;
        } else {
            divisor++;                     // 나누어 떨어지지 않는다면 나누는 수 1증가
        }
    } return true;
}

function reverse(str) {
    var reverStr = ''; // 빈문자열
    for (var i = str.length - 1; i >= 0 ; --i){
        reverStr = reverStr + str.charAt(i); //str의 i번째 문자를 빈 문자열에 추가
    } return reverStr;
}

자바스크립트 활용 - 소수 출력하기
while문과 if~else문을 사용하여, 소수를 판별하는 함수 isPrime의 코드를 작성해 봅시다! 소수란 약수가 1과 자기 자신밖에 없는 수입니다. 또, 1은 소수가 아닙니다.

document.write(isPrime(10)); //false 반환
document.write(isPrime(13)); //true 반환
Copy

지시사항
인자를 나누는 값인, 변수 divisor를 선언합니다. divisor의 초깃값은 2가 됩니다.

매개변수 n이 변수 divisor보다 큰 조건 하에서 while문이 작동되도록 합니다.

while문 안에 if else문을 삽입합니다.

만약 n을 divisor로 나눈 값의 나머지가 0이라면 false를 반환하고, 0이 아니라면, divisor++를 수행하도록 합니다.

while문의 조건이 false가 되면, true를 반환하도록 합니다.

단, while문을 나오고 true를 반환하기 전에 1은 소수가 아니기 때문에 n이 1일 경우 false를 반환하도록 합니다.

Tips
함수의 반환값이 true라면, 함수에 입력된 인자가 소수라는 의미입니다.

```python
'''
// 매개변수 n이 소수라면 true를, 소수가 아니라면 false를 반환합니다.
function isPrime(n){
    var divisor = 2;
    
    while ( divisor < n){
        if (n % divisor === 0) {
            return false;
        }else{
            divisor++;
        }
    }
    
    if (n === 1){
        return false;
    }
    return true;
    
}


// 아래 부분은 채점을 위해 작성되었습니다. 수정하지 마세요.
for(var i = 1; i <= 10; i++) {
    document.writeln(i, isPrime(i));
}
'''

'''
1false 2true 3true 4false 5true 6false 7true 8false 9false 10false
'''
```

자바스크립트 활용 - 문자열 거꾸로 출력하기
for문을 사용하여, 함수의 인자로 전달된 문자를 거꾸로 출력하는 함수 reverse의 코드를 작성해 봅시다!

document.write(reverse("Hello"));

//olleH 출력
Copy

지시사항
4줄에 for ( ) { } 을 입력하여 조건문 생성을 시작합니다.

for의 변수로 var i = str.length-1를 입력합니다.

for의 조건으로 i >= 0을 입력합니다.

for의 조건으로 i--을 입력합니다.

조건문 실행 시, 비어있는 변수 reverStr에 str.charAt(i)를 삽입하도록 합니다.

for문 실행을 완료하였다면, reverStr을 반환하도록 합니다.

Tips
str.length-1인 이유는, 데이터의 총 길이에서 1을 빼야, 해당 데이터의 인덱스(좌푯값)가 되기 때문입니다.

또한, 마지막 문자부터 거꾸로 출력해야 하므로, str.length-1번째부터 하나씩 감소해서, 0번째 문자까지 출력하는 것입니다.

```python
'''
function reverse(str){

    var reverStr = "";
    for (var i = str.length - 1; i >= 0; i--){
        reverStr = reverStr + str.charAt(i);
    }
    return reverStr;
}

document.write(reverse("Nice to meet you"));
'''

'''
uoy teem ot eciN
'''
```
