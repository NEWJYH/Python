```python
from selenium import webdriver


class Country:
    # ì§€ì‹œì‚¬í•­ 1ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # ì›¹ì— ìˆëŠ” ë°ì´í„°ë¥¼ êµ¬ì¡°í™”ëœ ë°ì´í„°(Structured Data)ë¡œ ë§Œë“¤ê¸° ìœ„í•´ 
    # class ë¥¼ ë¨¼ì € ì •ì˜í•©ë‹ˆë‹¤. 
    # ë©¤ë²„ ë³€ìˆ˜ë¡œ ë“¤ì–´ê°€ì•¼í•  ê²ƒì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.
    # êµ­ê°€ëª…
    # ìˆ˜ë„
    # ì¸êµ¬
    # ë©´ì 
    
    # ìƒì„±ì
    def __init__(self, name, capital, population, area):
        self.name = name
        self.capital = capital
        self.population = int(population)
        if 'E' in area:
            a, b = area.split('E') # a = 1.4 b = 7
            self.area = float(a) * (10 ** int(b))
        else:
            self.area = float(area)



with webdriver.Firefox() as driver:
    driver.get("https://www.scrapethissite.com/pages/simple/")

    # ì§€ì‹œì‚¬í•­ 2ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # êµ­ê°€ë³„ ì •ë³´ê°€ ë‹´ê¸´ ìš”ì†Œë¥¼ ëª¨ë‘ ê°€ì ¸ì˜¤ê³ , 
    # ê° ìš”ì†Œë¥¼ íŒŒì´ì¬ classì¸ Countryì˜ ì¸ìŠ¤í„´ìŠ¤ë¡œ ë§Œë“¤ì–´ 
    # country_list ì— ì¶”ê°€í•©ë‹ˆë‹¤.
    country_list = []
    div_list = driver.find_elements_by_class_name('country')
    for div in div_list:
        name = div.find_element_by_tag_name('h3').text
        capital = div.find_element_by_class_name('country-capital').text
        population = div.find_element_by_class_name('country-population').text
        area = div.find_element_by_class_name('country-area').text
        # ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
        country = Country(name , capital, population, area)
        # ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
        country_list.append(country)
    
    # ì§€ì‹œì‚¬í•­ 3ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # ëª¨ë“  êµ­ê°€ì˜ ìˆ˜ë„ë§Œ ë”°ë¡œ list ë¥¼ ë§Œë“­ë‹ˆë‹¤. 
    # ì´ ìˆ˜ë„ ëª©ë¡ì„ sort() ë˜ëŠ” sorted() ë¥¼ ì´ìš©í•˜ì—¬ ì‚¬ì „ ìˆœìœ¼ë¡œ ì •ë ¬í•˜ê³ ,
    # ëª©ë¡ì˜ 30ë²ˆì§¸ ì›ì†Œë¥¼ ì°¾ì•„ ì¶œë ¥í•©ë‹ˆë‹¤.
    capital_list = []
    for country in country_list:
        capital_list.append(country.capital)
    
    capital_list.sort()
    print(capital_list[29])
    # ì§€ì‹œì‚¬í•­ 4ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # í”íˆë“¤ 60ì–µ ì§€êµ¬ì´Œì´ë¼ëŠ”ë°, 
    # ì´ ë°ì´í„°ì—ì„  ê³¼ì—° ì–´ë–¨ê¹Œìš”? 
    # ëª¨ë“  êµ­ê°€ì˜ ì¸êµ¬ë¥¼ sum() ì„ ì´ìš©í•˜ì—¬ ë”í•´ì„œ ì¶œë ¥í•©ë‹ˆë‹¤.
    pop_list = []
    
    for country in country_list:
        pop_list.append(country.population)

    print(sum(pop_list))


    population_sum  = 0
    for country in country_list:
        population_sum += country.population
    print(population_sum)
```

ê²€ìƒ‰ì–´ë¥¼ ì…ë ¥(send_keys())í•˜ê³  Search ë²„íŠ¼ì„ í´ë¦­(click())í•©ë‹ˆë‹¤. ê²€ìƒ‰ì–´ëŠ” New ì…ë‹ˆë‹¤.

Newë¡œ ê²€ìƒ‰í•˜ë©´ ì´ ì„¸ íŒ€ì´ ë‚˜ì˜¬í…ë°, ì—°ë„ë³„ ê° íŒ€ì˜ ê¸°ë¡ì„ Record ì¸ìŠ¤í„´ìŠ¤ë¡œ ë§Œë“¤ì–´ record_list ì— ì €ì¥í•©ë‹ˆë‹¤. ì•„ë§ˆë„ ì—¬ëŸ¬ pageê°€ ë‚˜ì˜¬í…ë°, ì•„ë˜ì˜ Tipsë¥¼ ì°¸ê³ í•˜ì—¬ ëª¨ë“  ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ìˆë„ë¡ í•©ë‹ˆë‹¤.

record_list ë¥¼ ì´ìš©í•˜ì—¬ ê° ì—°ë„ë³„ ì„¸ íŒ€ì´ ìŒ“ì€ ìŠ¹ìˆ˜ì˜ í•©ì„ êµ¬í•´ì„œ win_dictì— ë„£ìŠµë‹ˆë‹¤. í•´ë‹¹ ì‚¬ì „ì˜ ê° keyëŠ” ì—°ë„, valueëŠ” ìŠ¹ìˆ˜ì…ë‹ˆë‹¤.

ìŠ¹ìˆ˜ë¥¼ ê°€ì¥ ë§ì´ ìŒ“ì€ ì—°ë„ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.

Tips!

```python
from selenium import webdriver
from typing import NamedTuple # namedtuple

class Record(NamedTuple):
    # ì§€ì‹œì‚¬í•­ 1ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # ì›¹ì— ìˆëŠ” ë°ì´í„°ë¥¼ êµ¬ì¡°í™”ëœ ë°ì´í„°(Structured Data)ë¡œ ë§Œë“¤ê¸° ìœ„í•´
    # class ë¥¼ ë¨¼ì € ì •ì˜í•©ë‹ˆë‹¤. 
    # ë©¤ë²„ ë³€ìˆ˜ë¡œ ë“¤ì–´ê°€ì•¼í•  ê²ƒì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.
    # íŒ€ëª…
    # ê¸°ë¡ì—°ë„
    # ìŠ¹ìˆ˜
    # íŒ¨ìˆ˜
    name: str
    year: int
    wins: int
    losses: int
    


with webdriver.Firefox() as driver:
    driver.get("https://www.scrapethissite.com/pages/forms/")
    # ì§€ì‹œì‚¬í•­ 2ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # ê²€ìƒ‰ ê¸°ëŠ¥ì„ í™œìš©í•˜ê¸° ìœ„í•´, 
    # ë‹¨ì–´ë¥¼ ì…ë ¥í•  ìš”ì†Œì™€ Search ë²„íŠ¼ ìš”ì†Œë¥¼ ì°¾ìŠµë‹ˆë‹¤.
    # idëŠ” í•˜ë‚˜ì˜ htmlë¬¸ì„œì—ì„œ ê³ ìœ í•œ ê°’ì„ ê°€ì§„ë‹¤ ê·¸ë ‡ê²Œ ì§œëŠ”ê²Œ ì•½ì†
    input_e = driver.find_element_by_id('q')
    # idê°€ ì—†ëŠ” ê²½ìš° classë¡œë„ ì°¾ì„ ìˆ˜ ìˆì§€ë§Œ
    # ë°˜ë“œì‹œ í•˜ë‚˜ë§Œ ìˆì„ ê²½ìš°ë§Œ ìˆëŠ”ê²ƒì´ ì•„ë‹ˆê¸°ë•Œë¬¸ì—
    # xpathë¡œ ì°¾ì•„ì¤€ë‹¤.
    search_e = driver.find_element_by_xpath(
    '//*[@id="hockey"]/div/div[4]/div/form/input[2]')

    # ì§€ì‹œì‚¬í•­ 3ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # ê²€ìƒ‰ì–´ë¥¼ ì…ë ¥(send_keys())í•˜ê³  
    # Search ë²„íŠ¼ì„ í´ë¦­(click())í•©ë‹ˆë‹¤. ê²€ìƒ‰ì–´ëŠ” New ì…ë‹ˆë‹¤.
    input_e.send_keys("New")
    search_e.click()
    
    # ì§€ì‹œì‚¬í•­ 4ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # Newë¡œ ê²€ìƒ‰í•˜ë©´ ì´ ì„¸ íŒ€ì´ ë‚˜ì˜¬í…ë°, 
    # ì—°ë„ë³„ ê° íŒ€ì˜ ê¸°ë¡ì„ Record ì¸ìŠ¤í„´ìŠ¤ë¡œ ë§Œë“¤ì–´
    # record_list ì— ì €ì¥í•©ë‹ˆë‹¤. 
    # ì•„ë§ˆë„ ì—¬ëŸ¬ pageê°€ ë‚˜ì˜¬í…ë°, 
    # ì•„ë˜ì˜ Tipsë¥¼ ì°¸ê³ í•˜ì—¬ ëª¨ë“  ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ìˆë„ë¡ í•©ë‹ˆë‹¤.
    record_list = []
    ul = driver.find_element_by_class_name("pagination")
    a_list = ul.find_elements_by_tag_name("a")
    
    url_list = []
    # ë‹¤ìŒí˜ì´ì§€ë¡œ ê°€ëŠ” ë§í¬ ì œê±°(ë§ˆì§€ë§‰ ì¸ë±ìŠ¤)
    for a in a_list[:-1]:
        url_list.append(a.get_attribute('href'))
    #print(url_list)
    
    for url in url_list:
        driver.get(url)
        
        tbody = driver.find_element_by_tag_name('tbody')
        team_list = tbody.find_elements_by_class_name('team')
        
        for team in team_list:
            name = team.find_element_by_class_name('name').text
            year = team.find_element_by_class_name('year').text
            wins = team.find_element_by_class_name('wins').text
            losses = team.find_element_by_class_name('losses').text
            record = Record(name = name,
                            year = int(year),
                            wins = int(wins),
                            losses = int(losses))
            #print(record)
            record_list.append(record)
        
    # ì§€ì‹œì‚¬í•­ 5ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # record_list ë¥¼ ì´ìš©í•˜ì—¬ 
    # ê° ì—°ë„ë³„ ì„¸ íŒ€ì´ ìŒ“ì€ ìŠ¹ìˆ˜ì˜ í•©ì„ êµ¬í•´ì„œ win_dictì— ë„£ìŠµë‹ˆë‹¤. 
    # í•´ë‹¹ ì‚¬ì „ì˜ ê° keyëŠ” ì—°ë„, valueëŠ” ìŠ¹ìˆ˜ì…ë‹ˆë‹¤.
    win_dict = {}  # {1990: 100, 1991: 110, 1992: 120, ...}
    for record in record_list:
        if record.year not in win_dict:
            win_dict[record.year] = 0
        win_dict[record.year] += record.wins
    #print(win_dict)
    # ì§€ì‹œì‚¬í•­ 6ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # ìŠ¹ìˆ˜ë¥¼ ê°€ì¥ ë§ì´ ìŒ“ì€ ì—°ë„ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
    best_year = 1990
    for year in win_dict:
        if win_dict[year] > win_dict[best_year]:
            best_year = year
        
    print(best_year)
    # print(win_dict[best_year])
```

```python
from typing import NamedTuple
from selenium import webdriver
import time

class Film(NamedTuple):
    # ì§€ì‹œì‚¬í•­ 1ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # ë°ì´í„°(Structured Data)ë¡œ ë§Œë“¤ê¸° ìœ„í•´ 
    # class ë¥¼ ë¨¼ì € ì •ì˜í•©ë‹ˆë‹¤. 
    # ë©¤ë²„ ë³€ìˆ˜ë¡œ ë“¤ì–´ê°€ì•¼í•  ê²ƒì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.
    # ì˜í™” ì œëª©
    # í•´ë‹¹ë…„ë„
    # í›„ë³´ë“±ë¡ ìˆ˜
    # ìˆ˜ìƒ ê°œìˆ˜
    name : str
    year : int
    nominations : int
    awards : int



with webdriver.Chrome() as driver:
    driver.get("https://www.scrapethissite.com/pages/ajax-javascript/")

    # ì§€ì‹œì‚¬í•­ 2ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # ê° ì—°ë„ë³„ ë§í¬ ìš”ì†Œë¥¼ ì°¾ìŠµë‹ˆë‹¤.
    a_list = driver.find_elements_by_class_name('year-link')
    
    # ì§€ì‹œì‚¬í•­ 3ë²ˆê³¼ 4ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    film_list = list()
    # ê° ì—°ë„ë³„ ë§í¬ ìš”ì†Œë¥¼ í•˜ë‚˜ì”© í´ë¦­í•´ê°€ë©° 
    # ëª¨ë“  ì˜í™” ë°ì´í„°ë¥¼ ì¶”ì¶œí•´ì„œ Film ì¸ìŠ¤í„´ìŠ¤ë¡œ ë§Œë“¤ì–´ 
    # film_list ì— ì¶”ê°€í•©ë‹ˆë‹¤.
    for a in a_list:
        a.click()
        # ê·¸ëŸ°ë° ë²„íŠ¼ì„ í´ë¦­í•˜ë©´ 
        # ë°ì´í„°ë¥¼ ì„œë²„ì—ì„œ ê°€ì ¸ì˜¤ëŠ” ajax í†µì‹ ì´ ì´ë£¨ì–´ì§‘ë‹ˆë‹¤. 
        # ë•Œë¬¸ì— ë¡œë”©ì— ì‹œê°„ì´ ê±¸ë¦¬ê²Œ ë˜ì–´ 
        # Wait ê¸°ëŠ¥ì´ í•„ìˆ˜ì ì…ë‹ˆë‹¤. 
        # ì´ ì‹¤ìŠµì— ì ì ˆí•œ Wait ê¸°ëŠ¥ì„ ì¶”ê°€í•´ì„œ ë°ì´í„° ì¶”ì¶œì„ í•´ë´…ì‹œë‹¤.
        time.sleep(5)
        tbody = driver.find_element_by_tag_name('tbody')
        tr_list = tbody.find_elements_by_tag_name('tr')
        
        for tr in tr_list:
            film_list.append(Film
                (name = tr.find_element_by_class_name('film-title').text,
                 year = int(a.text),
                 nominations = int(tr.find_element_by_class_name('film-nominations').text),
                 awards=int(tr.find_element_by_class_name('film-awards').text))
                 )
                 
        
    # ì§€ì‹œì‚¬í•­ 5ë²ˆì„ ì‘ì„±í•˜ì„¸ìš”.
    # film_list ë¥¼ ì´ìš©í•´ì„œ, 
    # 10ê°œ ì´ìƒ í›„ë³´ì— ì˜¬ëë˜ ì‘í’ˆì˜ ì‘í’ˆëª…ì„ 
    # ì‚¬ì „ìˆœìœ¼ë¡œ ì •ë ¬í•˜ê³ , ëª¨ë‘ ì¶œë ¥í•´ë´…ì‹œë‹¤.
    many_nomination_list = list()
    for film in film_list:
        if film.nominations >= 10:
            many_nomination_list.append(film.name)
            
    print(sorted(many_nomination_list))
```
