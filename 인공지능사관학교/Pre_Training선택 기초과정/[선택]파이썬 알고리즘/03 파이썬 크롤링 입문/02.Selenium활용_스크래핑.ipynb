지시사항
브라우저를 실행해봅시다. 브라우저는 Firefox 를 사용하며, with 문을 활용해서 코드를 작성해봅시다.

index.html 을 전달하고 있는 서버에 접속합니다. 서버의 주소는 http://localhost:8080 입니다.

웹 사이트의 제목을 나타내는 title 요소를 추출하고, 그 요소의 text 를 출력해봅시다.

이미지를 나타내는 img 요소를 모두 추출하고, 각 요소의 src 속성의 속성값을 출력해봅시다.

모든 div 요소에 포함된 모든 p 요소를 추출하고, 각 요소의 text 를 출력해봅시다.

Tips!
속성값을 출력하는 방법은 다음과 같습니다.
e = driver.find_element...
print(e.get_attribute("src"))

```python
# index.html
'''
<!DOCTYPE html>
<html>
  <head>
    <style type="text/css">
      img {
        width: 80px;
      }
    </style>
    <meta charset="UTF-8" />
    <title>Selenium 실습</title>
  </head>

  <body>
    <img src="img/js01.png" alt="JS logo" />
    <p>JavaScript</p>

    <div>
      <p>Text</p>
    </div>

    <p>No image</p>

    <img src="img/css03.png" alt="CSS logo" />
    <p>CSS</p>

    <div>
      <img src="img/html03.png" alt="HTML logo" />
      <p>HTML</p>
    </div>
  </body>
</html>

'''
```

```python
from selenium import webdriver

# 지시사항 1번을 작성하세요.
# 브라우저를 실행해봅시다.
# 브라우저는 Firefox를 사용하며, with문을 활용해서 코드를 작성해봅시다.
with webdriver.Firefox() as driver:
    
    # 지시사항 2번을 작성하세요.
    # index.html을 전달하고 있는 서버에 접속합니다.
    # 서버의 주소는 http://localhost:8080입니다.
    driver.get("http://localhost:8080")
    
    # 지시사항 3번을 작성하세요.
    # 웹사이트의 제목을 나타내는 title 요소를 추출하고, 
    # 그 요소의 text를 출력해봅시다.
    title = driver.find_element_by_tag_name('title')
    print(title.text)
    
    # 지시사항 4번을 작성하세요.
    # 이미지를 나타내는 img요소를 모두 추출하고,
    # 각 요소의 src속성의 속성값을 출력해봅시다.
    img_list = driver.find_elements_by_tag_name('img')
    for img in img_list:
        print(img.get_attribute("src"))
    
    # 지시사항 5번을 작성하세요.
    # 모든 div 요소에 포함된 모든 p요소를 추출하고, 
    # 각 요소의 text를 출력해봅시다.
    div_list = driver.find_elements_by_tag_name('div')
    for div in div_list:
        p_list = div.find_elements_by_tag_name('p')
        
        for p in p_list:
            print(p.text)
```

```python
# index.html
'''
<!DOCTYPE html>
<html>
  <head>
    <style>
      .bold {
        font-weight: bold;
      }
      .big {
        font-size: 24px;
      }
    </style>
    <meta charset="UTF-8" />
    <title>Selenium 실습</title>
  </head>

  <body>
    <h1>쿠지라이식 라면 레시피</h1>

    <h2>준비물</h2>
    <ul>
      <li class="big">라면 1봉지</li>
      <li>물 300ml</li>
      <li>계란 1개</li>
      <li class="bold">치즈 1/2개</li>
      <li>김가루와 쪽파 조금</li>
    </ul>

    <h2>조리법</h2>
    <ol>
      <li>물과 <span class="big bold">스프 2/3봉지</span>를 넣고 끓인다.</li>
      <li>면을 넣어 반쯤 익힌다.</li>
      <li>계란을 넣고, 뚜껑을 덮어 반숙으로 익힌다.</li>
      <li class="bold">치즈, 김가루, 쪽파 등 고명을 올려 먹는다.</li>
    </ol>
  </body>
</html>
'''
```

```python
from selenium import webdriver

# 지시사항 1번을 작성하세요.
# Selenium 설정은 다음과 같습니다.
# 1. 브라우저 : Firefox
# 2. 접속 url : http://localhost:8080
with webdriver.Firefox() as driver:
    driver.get('http://localhost:8080')
    # 지시사항 2번을 작성하세요.
    # ol 태그 안에 있는 li 태그를 가진 모든 요소를 추출하고,
    # 각 요소의 text를 출력해봅시다.
    ol = driver.find_element_by_tag_name('ol')
    li_list = ol.find_elements_by_tag_name('li')
    for li in li_list:
        print(li.text)
    # 지시사항 3번을 작성하세요.
    # big class가 적용된 모든 요소를 추출하고,
    # 각 요소의 text를 출력해봅시다.
    big_list = driver.find_elements_by_class_name('big')
    for big in big_list:
        print(big.text)
    
    # 지시사항 4번을 작성하세요.
    # ul 태그 안에 있는 요소 중, 
    # bold class가 적용된 요소를 추출하고,
    # 그 요소의 text를 출력해봅시다.
    ul = driver.find_element_by_tag_name('ul')
    bold = ul.find_element_by_class_name('bold')
    print(bold.text)
    # ul_list = driver.find_elements_by_tag_name('ul')
    # for ul in ul_list:
    #     bold = ul.find_elements_by_class_name('bold')
    #     for i in bold:
    #         print(i.text)
```

위의 xpath에 해당하는 요소를 찾아 그 요소의 text 를 출력해봅시다.

Tips!
xpath를 이용해 요소를 찾을 때, 해당 class가 포함되어 있는지 확인하고 싶을 때는 contains()를 사용합니다.

'//li[contains(@class, "big")]'
#li 중, class 속성에 "big"이 포함되어 있는 경우

```python
'''
<!DOCTYPE html>
<html>
  <head>
    <style>
      .bold {
        font-weight: bold;
      }
      .big {
        font-size: 24px;
      }
    </style>
    <meta charset="UTF-8" />
    <title>Selenium 실습</title>
  </head>

  <body>
    <h1>쿠지라이식 라면 레시피</h1>

    <h2>준비물</h2>
    <ul>
      <li class="big">라면 1봉지</li>
      <li>물 300ml</li>
      <li>계란 1개</li>
      <li class="bold">치즈 1/2개</li>
      <li>김가루와 쪽파 조금</li>
    </ul>

    <h2>조리법</h2>
    <ol>
      <li>물과 <span class="big bold">스프 2/3봉지</span>를 넣고 끓인다.</li>
      <li>면을 넣어 반쯤 익힌다.</li>
      <li>계란을 넣고, 뚜껑을 덮어 반숙으로 익힌다.</li>
      <li class="bold">치즈, 김가루, 쪽파 등 고명을 올려 먹는다.</li>
    </ol>
  </body>
</html>

'''
```

```python
from selenium import webdriver

# 지시사항 1번을 작성하세요.
# selenium설정

with webdriver.Firefox() as driver:
    driver.get("http://localhost:8080")
    
    # 지시사항 2번을 작성하세요.
    # ol태그 안에 있는 li 태그를 가진 모든 요소를 나타내는 xpath를
    # xpath변수에 할당하고 출력
    xpath = '/html/body/ol//li'
    print(xpath)

    # 지시사항 3번을 작성하세요.
    # 위의 xpath에 해당하는 모든 요소를 찾아 각 요소의 text 를 출력해봅시다.
    li_list = driver.find_elements_by_xpath(xpath)
    for li in li_list:
        print(li.text)
    # 지시사항 4번을 작성하세요.
    # big class가 적용된 모든 요소를 나타내는 xpath를 
    # xpath2 변수에 할당하고 출력해봅시다.
    xpath2 = '//*[contains(@class, "big")]'
    print(xpath2)

    # 지시사항 5번을 작성하세요.
    # 위의 xpath에 해당하는 모든 요소를 찾아 각 요소의 text 를 출력해봅시다.
    big_list = driver.find_elements_by_xpath(xpath2)
    for big in big_list:
        print(big.text)
    
    # 지시사항 6번을 작성하세요.
    # ul 태그 안에 있는 요소 중, 
    # bold class가 적용된 요소를 나타내는 xpath를 
    # xpath3 변수에 할당하고 출력해봅시다.
    xpath3 = '//ul//*[contains(@class, "bold")]'
    print(xpath3)

    # 지시사항 7번을 작성하세요.
    # 위의 xpath에 해당하는 요소를 찾아 그 요소의 text 를 출력해봅시다.
    bold_list = driver.find_elements_by_xpath(xpath3)
    for bold in bold_list:
        print(bold.text)
```
