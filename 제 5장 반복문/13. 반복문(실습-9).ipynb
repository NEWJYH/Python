```python
# 더블루프를 이용하여 아래와 같이 출력하는 프로그램을 작성하시오.
# 출력 결과
#    *     # 1
#   ***    # 3
#  *****   # 5
# *******  # 7
#********* # 9

print("double loop")
for y in range(1,6):
    for x in range(5-y):
        print(" ",end="")
    for z in range(2*y-1):
        print("*",end="")
    print()

print("foramt")
for i in range(1,6):
    print("{0}{1}".format((" "*(5-i)),("*"*(2*i-1))))
    # 중앙정렬을 위해서는 ^ 특수문자를 사용하면 된다.
print("^사용")
for i in range(1,11,2):
    print("{:^10}".format("*"*i))
    # 중앙정렬을 위해서는 ^ 특수문자를 사용하면 된다.
    
    
print("F스트링")
for i in range(1,6):
    print(f"{' '*(5-i)}{'*'*(2*i-1)}")
```

```python
# 소수를 2부터 계속 더할 때, 2000까지 루프를 돌리고
# 소수와 합계
# 마지막으로 더해지는 소수는 얼마 인지 출력하는 프로그램을 작성해 보자.
#(더블루프, 조건식 사용

sum = 0

for num in range(2, 2001):
    for start_num in range(2, num +1):
        if num % start_num == 0:
            break
    if num == start_num:
        print("소수 : ", start_num)
        sum += start_num
        print("합계 : ", sum)
        lastData = start_num
        print("마지막 소수의 값 : ",lastData)
        print("==========================")
```
