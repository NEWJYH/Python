```python
# 사용자로부터 출력하고 싶은 구구단을 출력하는 프로그램을 만들기

INPUT_NUM = int(input("출력하고 싶은 구구단은 ? : "))

for i in range(1, 10):
    print(f"{INPUT_NUM} * {i} = {INPUT_NUM * i:>3}")
```

```python
# 사용자로부터 2개의 정수 값을 입력받고
# 첫 번째 입력받은 값부터 두 번째 입력 받은 값까지의 범위에서
# 3의 배수이면서 4의 배수인 수를 제외하고 출력하는 프로그램을 작성하시오.

n1 = int(input("첫수 : "))
n2 = int(input("두번쨰 수 : "))

for i in range(n1, n2+1):
    if i%3 == 0 and i%4 == 0 :
        continue
        # continue는 조건식이 참이면 아래 문장을 실행치 않고
        # fot문으로 이동하여 증가한다.
        
    print(f"{i}",end=" ")
```

```python
# 터틀그래픽 라이브러리를 이용해서 별 그려보기
import turtle

t = turtle.Pen()

for i in range(5):
    t.forward(50)
    t.right(144)
    
# 터틀그래픽 창이 클릭이 되여야 화면에서 사라지게 하는 코드
turtle.exitonclick()
```

```python
# 터틀 그래픽을 이용하여 사각형 3개를 그려보자.
# 단 조건은 사각형을 20도씩 기울여 저 있다.

import turtle

t = turtle.Pen()

for i in range(15):
    t.left(20) 
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.left(90)
    t.forward(50)
    t.left(90)

turtle.exitonclick()
```
