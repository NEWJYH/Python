```python
# 사용자로부터 임의의 갯수의 성적을 입력받아서
# 합계와 평균을 계산한 후 
# 출력하는 프로그램을 작성해보시오.
# 단, 센티널은 음수의 값을 사용하시오.

print("총점과 평균을 구하는 프로그램 ")
print("음수를 입력하면 종료")
sum = 0
count = 0
average = 0
i = 0
while i >= 0 :
    score = int(input("너의 점수를 입력해주세요 : "))
    if score >= 0 :
        sum += score
        count += 1
    else:
        average = sum / count
        i = -1 
print(f"입력받은 과목 {count} 총점은 {sum} 종합 평균은 {average:>3,.2f}")
```

```python
# 임의의 숫자를 발생시켜 숫자를 맞추는 게임을 만들어보기

from random import *

cnt = 0
num = randint(1,100)
print("발생난수의 값 :",num)

print("1부터 100사이의 숫자를 맞추어 보세요")
print("기회는 10번입니다.")
while cnt < 10:
    guess = int(input("숫자를 입력하세요 : \n\n"))
    cnt += 1
    if guess < num :
        print("입력한 수가 난수보다 낮습니다.")
    elif guess > num:
        print("입력한 수가 난수보다 높습니다.")
    elif guess == num :
        print("정답입니다.")
        print(f"입력한수 {guess} 정답 {num} 시도횟수 {cnt} ")
        ExitorNew = input("게임을 종료하시겠습니까? (y,n) :")
        if ExitorNew == "y":
            break
        elif ExitorNew == "n":
            cnt = 0
            num = randint(1,100)
            print("새로운 게임을 다시 시작합니다.")
    
    print(f"남은 도전 횟수 : ",10 - cnt)
print("게임이 종료 되었습니다.")
```
