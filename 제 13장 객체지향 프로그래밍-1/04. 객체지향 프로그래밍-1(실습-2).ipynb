```python
# 클래스 Tv를 설계하는 코드
class Tv:
    # 맴버변수(속성,필드)선언
    color = "" # 색상
    Power = False # 전원
    channel = 0 # 채널
    volume = 0 # 볼륨
    
    # Tv의 전원을 켜거나 끄는 맴버 메서드
    def power(self, power):
        self.Power = power
        if self.Power == True:
            print("Tv 전원을 켰습니다")
        else:
            print("Tv 전원을 껏습니다")
            
        
    # Tv의 채널을 올리고 내리는 기능을 하는 멤버 메서드
    def channelUp(self,channel):
        if self.Power == False:
            return
        if channel < 0:
            print("채널은 음수 일수 없습니다.")
            return
        self.channel += channel
    def channelDown(self,channel):
        if self.Power == False:
            return
        if channel < 0:
            print("채널은 음수 일수 없습니다.")
            return
        self.channel -= channel
    def channelSet(self,channel):
         if self.Power == False:
            return
         if channel < 0:
            print("채널은 음수 일수 없습니다.")
            return
         self.channel = channel
        
    # Tv의 볼륨을 올리고 내리는 기능을 하는 메서드
    def volumeUp(self,volume):
        if self.Power == False:
            return
        if volume < 0:
            print("볼륨은 음수 일수 없습니다.")
            return
        self.volume += volume
    def volumeDown(self,volume):
        if self.Power == False:
            return
        if volume < 0:
            print("볼륨은 음수 일수 없습니다.")
            return
        self.volume -= volume
        if self.volume <0:
            self.volume = 0
    def volumeSet(self,volume):
        if self.Power == False:
            return
        if volume < 0:
            print("볼륨은 음수 일수 없습니다.")
            return
        self.volume = volume
        
    # Tv셋팅을 확인할수 있는 메서드
    # __str__() 인스턴스의 멤버 변수의 값을 찍어보는 용도로 사용된다. 
    def __str__(self):
        print(f"Tv의 전원 : {self.Power}")
        print(f"Tv의 색상 : {self.color}")
        print(f"Tv의 채널 : {self.channel}")
        print(f"Tv의 볼륨 : {self.volume}")                
```

```python
# from Tv import *
if __name__=="__main__":
    # 인스턴스 생성
    tv1 = Tv()
    tv2 = Tv()
    # 인스턴스의 필드와 메서드를 tv1 인스턴스명(참조변수) 조작을 하고있다.
    tv1.color = "검정색"
    tv1.power(True)
    tv1.channelSet(45)
    tv1.volumeSet(25)
    tv1.__str__()
    tv1.power(False)
    tv1.__str__()
    print("---------------")
    tv2.color = "파란색"
    tv2.power(True)
    tv2.channelSet(45)
    tv2.volumeSet(25)
    tv2.__str__()
    tv2.power(False)
    tv2.__str__()
```
