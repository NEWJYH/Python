```python
from tkinter import *

# 라벨 텍스트 설정
labeltxt = "mytkinter"

# 버튼 텍스트 설정
buttxt = f"버튼 클릭 카운트"
# 사용할 전역변수
num = 0
import random
# 버튼 콜백 함수
def play():
    global num
    num += 1
    colorindex = random.randrange(0, 5)
    color = ["red","green","orange","yellow","blue"]
    button["text"] = f"{num}번 클릭"
    button["bg"] = f"{color[colorindex]}"
    button["fg"] = f"{color[colorindex-1]}"

# 윈도우 생성
window  = Tk()
window.title("tkinter실습프로그램") # 타이틀 이름설정
window.geometry("1920x1080") # 사이즈 설정
window.option_add("*Font","궁서 25") # 옵션추가

# 라벨 설정
label = Label(window, text = labeltxt)
# 라벨 팩킹
label.pack() 

# 버튼 설정
button = Button(window,# 대상 윈도우
               text = buttxt, # 버튼안 글자
                bg = "yellow", # 버튼 안 배경색
                fg = "blue", # 버튼 안 글자색
                width = 80, # 버튼 너비
                height = 20, # 버튼 높이
                font = "고딕 50", # 고딕 50픽셀 설정
                command = play) # 버튼 클릭 콜백 함수

# 버튼 팩킹
button.pack(side = TOP, # 위쪽에 생성
            padx = 10)  # 왼쪽 여백 10픽셀

window.mainloop()
```

```python
# 배치관리자(Layout Manager) : 위젯들을 윈도우에 배치하는 방법,
# 기존값은 TOP 수직으로 쌓는다.
from tkinter import *

window = Tk()
window.title("배치관리자") # 윈도우 타이틀 설정 
# window.geometry("1920x1080") # 사이즈 설정
window.option_add("*Font","궁서 10") # 기본옵션추가 궁서체 10포인트

label = Label(window, # 라벨 대상 윈도우
             text = "배치관리자 실습") # 라벨 텍스트

label.pack() # 라벨 팩킹 윈도우 설정 완료

# 버튼 클릭 이벤트 콜백 함수 생성
# 사용할 전역변수
num = 0
num1 = 0
import random
# 버튼 콜백 함수
def play1():
    global num
    num += 1
    colorindex = random.randrange(0, 5)
    color = ["red","green","orange","yellow","blue"]
    button1["text"] = f"{num}번 클릭"
    button1["bg"] = f"{color[colorindex]}"
    button1["fg"] = f"{color[colorindex-1]}"

def play2():
    global num1
    num1 += 1
    colorindex = random.randrange(0, 5)
    color = ["red","green","orange","yellow","blue"]
    button2["text"] = f"{num1}번 클릭"
    button2["bg"] = f"{color[colorindex]}"
    button2["fg"] = f"{color[colorindex-1]}"


button1 = Button(window, # 대상 윈도우
                text = "버튼1",
                width = 30,
                height = 10,
                font = "고딕 20",
                command = play1)
button2 = Button(window, # 대상 윈도우
                text = "버튼2",
                width = 30,
                height = 10,
                font = "고딕 20",
                command = play2)
button3 = Button(window, # 대상 윈도우
                text = "버튼3",
                bg = "blue",
                width = 30,
                height = 10,
                font = "고딕 20")
# pack() 메서드 호출시 매개변수가 없는 것으로 
# 호출하면 수직으로 정렬이 이루어진다.
# 그 이유는 side 값이 기본적으로 TOP으로 설정이 되어 있기 때문이다.
# button1.pack()
# button2.pack()
button1.pack(side = LEFT,padx = 20,pady = 20)
button2.pack(side = LEFT,padx = 20,pady = 20)
button3.pack(side = LEFT,padx = 20,pady = 20)
# side 매개변수의 값은 TOP, LEFT, RIGHT, BOTTOM 존재하므로
# 배치할떄 적절히 이용하면 된다.
# padx는 가로로 안쪽 여백을 준다.
# pady는 세로로 안쪽 여백을 준다.
# 패안마바(패딩은 안쪽 여백, 마진은 바깥쪽 여백)

# 버튼의 text 속성을 바꾸는 방법
button1["text"] = "One"
button2["text"] = "Two"
button3["text"] = "Three"

window.mainloop()
```
