```python
# 색상에 대한 실습
from tkinter import *

window = Tk()
window.title("색상 연습")
window.geometry("800x600")
button = Button(window, text = "버튼을 클릭하세요")
button.pack()
# 속성값의 변경을 통하여 버튼의 배경색, 전경색을 바꾸는 방식
# 1. 실제 색상 원색 대입
# button["fg"] = "yellow"
# button["bg"] = "green"

# 2. 16진수를 이용하여 색상을 변경함
# 구글 검색 컬러피커
# button["fg"] = "#4287f5"
# button["bg"] = "#254194"

# 3. tkinter.colorchooser 모듈의 askcolor() 를 호출하면
# 대화상자를 통해서 색상의 값을 입력 받을 수 있다.
# askcolor() 메서드는 튜플을 리턴함으로 해당하는 16진수의 값이 인덱스로 1번째
# 에 존재하므로 아래와 같이 설정할 수 있다.
from tkinter.colorchooser import *
# from tkinter import colorchooser 
color = askcolor()
color1 = askcolor()
button["fg"] = color[1]
button["bg"] = color1[1]

window.mainloop()
```

```python
# 폰트에 대한 실습
# 모든 위젯이 font라는 속성을 거의 다 가지고 있다.
# 이 font의 속성을 변경하면
# 원하는 폰트로 변경시킬수가 있다.
# 폰트는 미리 객체로 생성해 놓고 여러 곳에서 사용하면 메모리가 절약된다.

from tkinter import *
from tkinter import font
class App:
    def __init__(self):
        self.root = Tk()
        self.root.geometry("500x600")
        #폰트 지정
        self.customFont = font.Font(family = "Helvetica", size = 12)
        bigger = Button(self.root, text = "폰트를 크게",command = self.BigFont)
        smaller = Button(self.root, text ="폰트를 작게", command = self.SmallFont)
        bigger.pack()
        smaller.pack()
        label = Label(self.root, text = "Hello world", font = self.customFont)
        label.pack()
        self.root.mainloop()
        
    def BigFont(self):
        size = self.customFont["size"]
        print(size)
        self.customFont.configure(size = size + 2)
        print(self.customFont)
        
    def SmallFont(self):
        size = self.customFont["size"]
        print(size)
        # 폰트의 속성을 변경하고자 하면 configure() 사용하면 된다.
        self.customFont.configure(size = size - 2)
        print(self.customFont)

        
if __name__ == "__main__":
    App()
```
